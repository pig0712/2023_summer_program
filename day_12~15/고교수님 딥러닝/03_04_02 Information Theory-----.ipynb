# Information Theory

## 1. 정보(Information)
$\quad$ **[정의]** Shannon의 information $I$ 는 다음과 같이 정의 됨<br>
>$$
I(y) = \log_2 \left( \frac{1}{p(y)} \right)
$$
> $\qquad$ 여기서 $p(y)$는 $y$가 발생할 확률

### 1.1 정보(Information)의 의미
- 어떤 사건의 발생 확률이 매우 작으면 정보량은 매우 크고 확률이 크면 정보량은 작다는 의미임

- 먼저 확률 $p(y)$와 정보 $I(y)$의 관계를 Graph로 그려보자 

```python
import matplotlib.pyplot as plt
import numpy as np

p_y = np.linspace(1e-5, 1, 100) 
I_y =np.log(1/p_y)
```

```python
plt.plot(p_y, I_y)
plt.axvline(x=0, color='k', zorder=0)
plt.axhline(y=0, color='k', zorder=0)
plt.text(1, -2, r'$p(y)$', color='#1190FF',  size=12)
plt.text(-0.15, 11, r'$I(y)$', color='#1190FF',  size=12)
plt.show()
```

$\Longrightarrow$ 정보량은 항상 0 또는양의 값(Non-negative)을 가진다<br><br>

$\Longrightarrow$ 확률이 작으면 정보량은 크다<br><br>

- "내일은 해가 서쪽에서 뜬다"라는 예를 생각해보면<br>
$ \ \ $ $\checkmark \ $  확률 $p$는 매우 작은 값이 되지만<br>
$ \ \ $ $\checkmark \ $  정보량 $I$은 매우 큰 값이 된다는 의미임


  구분 | $\qquad \qquad$ 설명
:---- | :----
<br> Symbols | <br> $$y_i, \quad i=1,..., K$$
<br> 각 symbol들이 발생할 확률 | <br> $$p_i = P(Y=y_i) \ \ge \ 0  \quad i=1, ..., K$$ <br> $$\sum_{i=1}^{K} p_i =1$$
<br> 각 symbol들의 정보량 | <br> $$I_i = log_2 \left( \frac{1}{p_i} \right), \quad i=1, ..., K$$

\begin{cases}

## 2. Entropy

### 2.1 Entropy의 정의
**[정의]** 확률변수 $Y$의 entropy는 Information의 수학적 기댓치로 정의 됨<br>
>$$
\begin{align}
H(Y) =& E_Y [I(Y)] \\
     =& E_Y \left[ log_2 \left( \frac{1}{P(Y)} \right) \right]  
\end{align}
$$

(i) 확률변수 $Y$가 이산확률변수인 경우

$$
H(Y) = \sum_y \left( \log_2 \frac{1}{P(Y=y)} \right) P(Y=y)
$$

$\quad$ 단 여기서 $ P(Y=y)$는 pmf(Probability Mass Function)임

(ii) 확률변수 $Y$가 연속확률변수인 경우

$$
H(Y) = \int_y \left( \log_2 \frac{1}{f(y)} \right) f(y)dy
$$

$\quad$ 단 여기서 $f(y)$는 pdf(Probability Density Function)임

 ### 2.2 Binary Entropy

- 다음과 같은 사건을 생각해보자 <br>
$ \ \ $ $\checkmark \ $ 시행의 결과가 성공, 실패 2가지로 나타나는 시행(Bernoulli Trial)에서 확률변수 $Y$를 다음과 같이 정의함 <br><br>

$$
Y =
\begin{cases} 
0, &    Y가 \ 성공인 \ 사건 \\\\
1, &    Y가 \ 실패인 \ 사건
\end{cases}
$$

- $P(Y=0) = p$라고 하면 $P(Y=1) = 1-p$임<br>

$$
\begin{align}
H(Y) =& E[I(Y)]  \\
     =& E \left[ log_2 \left( \frac{1}{P(Y)} \right) \right] \\
     =& \sum_{i=0}^{1} log_2 \left( \frac{1}{P(Y=i)} \right) P(Y=i) \\
     =& log_2 \left( \frac{1}{P(Y=0)} \right) P(Y=0) + log_2 \left( \frac{1}{P(Y=1)} \right) P(Y=1) \\
     =& log_2 \left( \frac{1}{p} \right) p + log_2 \left( \frac{1}{1-p} \right) (1-p) \\
     =& -plog_2 p -(1-p)log_2(1-p)
\end{align}
$$

### 예 3
Binary Entropy에서 $p_1 = p_2 = p = 1/2$라면<br><br>

$$
\begin{align}
H(Y) =& -plog_2 p -(1-p)log_2(1-p) \\
     =& - \frac{1}{2} log_2 \left( \frac{1}{2} \right) 
                           - \left( 1 - \frac{1}{2} \right) log_2 \left( 1 - \frac{1}{2} \right) \\
     =& \frac{1}{2} + \frac{1}{2} = 1
\end{align}                           
$$

### 2.4 결합확률분포의 Entropy

- 다음과 같이 2개의 확률변수 $X$, $Y$의 결합확률분포에 대한  Entropy로 확장이 가능

$$
\begin{align}
H(X,Y) =& E[I(X, Y)] \\ 
       =& \sum_{y} \sum_{x}  \log_2 \left( \frac{1}{P(x,y)} \right) P(X=x,Y=y)
\end{align}
$$

- 만일 $X$와 $Y$가 통계적 독립이라면 Entropy는 다음과 같음<br>
$ \ \ $ $\checkmark \ $ 확률변수 $X$와 $Y$가 통계적 독립  $ \iff $ $P(X=x, Y=y)=P(X=x) \times P(Y=y)$

$$
\begin{align}
H(X,Y) =& E[I(X, Y)] \\ 
       =& \sum_{y} \sum_{x} log_2 \left( \frac{1}{P(X=x, Y=y)} \right) P(X=x, Y=y) \\
       =& \sum_{y} \sum_{x} log_2 \left( \frac{1}{P(X=x)P(Y=y)} \right) P(X=x)P(Y=y) \quad (\because 통계적 \ 독립)\\
       =& \sum_{y} \sum_{x} \left(log_2 \left( \frac{1}{P(X=x)} \right) + log_2 \left( \frac{1}{P(Y=y)}                   \right) \right)P(X=x)P(Y=y) \quad (\because log의 성질)\\
       =& \sum_{y} \sum_{x} log_2 \left( \frac{1}{P(X=x)} \right) P(X=x)P(Y=y)
           +
          \sum_{y} \sum_{x}  log_2 \left( \frac{1}{P(Y=y)} \right) P(X=x)P(Y=y) \\
       =& \sum_{y} P(Y=y) \sum_{x} log_2 \left( \frac{1}{P(X=x)} \right) P(X=x)
           +
          \sum_{x} P(X=x) \sum_{y}  log_2 \left( \frac{1}{P(Y=y)} \right) P(Y=y) \\
       =& \sum_{y} P(Y=y) H(X) + \sum_{x} P(X=x) H(Y) \quad \\
       & \qquad \qquad (\because 확률의 \ 성질에 \ 의해, \ \sum_x P(X=x) = 1, \sum_y P(Y=y) = 1)\\
       =& H(X)+H(Y)
\end{align} 
$$

$\quad$ $\Longrightarrow$ 즉, 2개의 확률변수가 통계적 독립이면 결합 Entropy는 각각의 Entropy의 합과 같음

### 2.5 Conditional Entropy

**[Definition: Conditional Entropy]**<br>
>- $X$가 주어졌을 때 $Y$d]의 Conditional Entropy는 다음과 같이 정의 된다<br>
><br>
>$$
H_{p_{x, y}}(Y|X) = \sum_{x \in \mathcal{X}} H(Y|X=x)p(x)
$$

$$\begin{align} 
H_{p_{x, y}}(Y|X) =& \sum_{x \in \mathcal{X}} H(Y|X=x)p(x) \\
=& \sum_{x \in \mathcal{X}} p(x) \sum_{y \in \mathcal{Y}} log \left( \frac{1}{p(y|x)} \right) p(y|x) \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} p(x) p(y|x) \ log \left( \frac{1}{p(y|x)} \right)  \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} \ p(x, y) \ log \left( \frac{1}{p(y|x)} \right)  \\
\end{align}
$$

- Conditional Entropy와 Entropy의 관계

$$\begin{align} 
H_{p_{x, y}}(Y|X) =& \sum_{x \in \mathcal{X}} H(Y|X=x)p(x) \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} p(x, y) log \left( \frac{1}{p(y|x)} \right)  \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} p(x, y) log \left( \frac{p(x)}{p(x, y)} \right)  \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} p(x, y) log \left( \frac{1}{p(x, y)} \right)
 - \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} p(x, y) log \left( \frac{1}{p(x)} \right)  \\
=& H(X, Y) - H(X)  
\end{align}
$$

## 4. Mutual Information

### 4.1 Mutual Information 정의

**[Definition: Mutual Information]**<br>
>- $(X, Y)$를 space $ \left( \mathcal{X \times Y} \right)$에서 정의된 확률변수 쌍이라고 하자<br>
>- Joint probability distribution를 $P(X, Y)$라고 하고 maraginal probability distribution을 각각 $P(X)$와 $P(Y)$라고 하면 Mutual Information $I(X; Y)$은 다음과 같이 정의된다<br>
><br>
>$$\begin{align} 
I(X; Y) =& D_{KL} \left( P(X, Y) \| P(X)  P(Y) \right)  \\
        =& E_{P(X, Y)} \left[ log \left( \frac{P(X, Y)}{P(X)P(Y)} \right) \right] \\
        =& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( \frac{p(x,y)}{p(x)p(y)} \right) p(x, y)
\end{align}$$

- Mutual Information은 두 확률변수간의  상호 의존성을 측정하는 measure임

### 4.2 Mutual Information의 성질

#### (1) Nonnegativity
- Jensen's inequality를 이용하여 증명 가능 

$$
I(X; Y) \ge 0
$$

#### (2) Symmetry
- Jensen's inequality를 이용하여 증명 가능 

$$
I(X; Y) = I(Y; X)
$$

#### $\qquad$증명

$$
\begin{align}
I(X; Y) =& H(X) - H(X | Y) \\
        =& H(X) - H(Y | X) \\
        =& H(X) + H(Y) - H(X, Y) \\
        =& H(X, Y) - H(X|Y) - H(Y|X)
\end{align}
$$

#### (3) Mutual Information과 Entropy의 관계

$$
\begin{align}
I(X ; Y) 
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( \frac{p(x,y)}{p(x)p(y)} \right) p(x, y) \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( \frac{p(x,y)}{p(x)} \right) p(x, y) 
-
\sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( p(y) \right) p(x, y)  \\
=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( \frac{p(x,y)}{p(x)} \right) 
                                                     \left( p(x) \times \frac{p(x, y)}{p(x)} \right)
-
\sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( p(y) \right) p(x, y)  \\
=& \sum_{x \in \mathcal{X}} p(x) \sum_{y \in \mathcal{Y}} \left( - log \left( \frac{1}{p(y|x)} \right) p(y|x) \right) 
+ 
\sum_{y \in \mathcal{Y}} log \left( \frac{1}{p(y)} \right) \sum_{x \in \mathcal{X}} p(x, y)  \\
=& - \sum_{x \in \mathcal{X}} p(x) H(Y|X=x) 
+ H(Y) \\
=& - H(Y|X) + H(Y)
\end{align}  \\
$$

#### (4) Mutual Information과 Kullback Liebler Divergence의 관계

$$
\begin{align}
I(X; Y)=& \sum_{x \in \mathcal{X}} \sum_{y \in \mathcal{Y}} log \left( \frac{p(x,y)}{p(x)p(y)} \right) p(x, y) \\
=& \sum_{y \in \mathcal{Y}} p(y) \sum_{x \in \mathcal{X}}  log \left( \frac{1}{p(x)} \times \frac{p(x, y)}{p(y)} \right) \frac{p(x,y)}{p(y)} \\
=& \sum_{y \in \mathcal{Y}} p(y) \sum_{x \in \mathcal{X}} log \left( \frac{p(x|y)}{p(x)} \right) p(x|y) \\
=& \sum_{y \in \mathcal{Y}}  p(y) D_{KL} \left( p(x|y) \| p(x) \right)  \\
=& E_{Y}\left[ D_{KL} \left( p(x|y) \| p(x) \right) \right] \\
\end{align}
$$

## 5. Kullback-Leibler Divergence

### 5.1 Kullback-Leibler Divergence의 정의

(i) 두 확률분포의 차이를 계산하는 데에 사용하는 함수임<br>

(ii) 쿨백-라이블러 발산$ \left( Kullback–Leibler \ Divergence, KLD \ 또는\ D_{KL} \right) $의 정의

$\quad$ **[정의] Kullback-Leibler Divergence**<br>
>$\quad$ $\bullet$  두 확률변수 $Y$와 $\hat Y$에 대한 분포 $P$와  $Q$가 있을 때, 두 분포의 Kullback-Leibler Divergence는 <br> $\qquad$  확률분포 $Q$에  대한 확률분포 $P$의 Entropy임<br>
>
>$\qquad$ $\checkmark \ $ Discrete Random Variable인 경우
$$
\begin{align}
D_{KL}(P \| Q) =& E_{P} \left[ log_2 \left( \frac{P}{Q} \right) \right] \\
               =&\sum_x log_2 \left( \frac{P(x)}{Q(x)} \right) P(x) 
\end{align}
$$
>
>$\qquad$ $\checkmark \ $ Continuous Random Variable인 경우
$$
\begin{align}
D_{KL}(P \| Q) =& E_{P} \left[ log_2 \left( \frac{P}{Q} \right) \right] \\ 
               =& \int_{x \in \mathcal{X}} log_2 \left( \frac{P(x)}{Q(x)} \right) P(x) dx  
\end{align}
$$

$\qquad$ $\Longrightarrow \ $ 이상적인 확률분포 $P$에 대해, 그 분포를 근사하는 다른 분포 $Q$를 사용해 샘플링을 한다면 발생할 수 있는 <br> $\qquad \qquad \ \ $ information entropy 차이를 말함<br>

(iii) Kullback-Leibler Divergence은 비대칭으로, 두 값의 위치를 바꾸면 함수값도 달라짐 <br> 
$\quad$ $\checkmark \ $ 따라서 이 함수는 거리 함수는 아님

(iv) 기계학습에서는 다양한 거리 및 유사도를 측정할 수 있는 measure를 사용함<br>
$\quad$ [참고] 01_01_01 Data 간의 거리와 유사도 척도.ipynb<br>
$\quad$ $\checkmark \ $ $D_{KL}$은 상대 엔트로피(relative entropy), 정보 획득량(information gain), 인포메이션 다이버전스(information <br> $\qquad$ divergence)라고도 부름<br>
$\quad$ $\checkmark \ $정보이론에서는 상대 엔트로피, 기계학습의 결정 트리에서는 정보 획득량을 주로 사용함

### 5.2 KL-divergence의 특성
$\quad$ **특성 1**: KLD는 Non-negative임
$$
D_{KL}(P \| Q) \ge 0
$$

[증명]<br>

(i) 다음 그림과 같은 convex function $f(x)$를 생각함

(ii) convec function의 정의에 따라 $0 \le \alpha \le 1$에 대해 다음식이 성립함

$$
\begin{align}
f(\alpha x_1 + (1 - \alpha) x_2) \le& \alpha f(x_1) + (1-\alpha) f(x_2) \quad (Jensen's \ \ inequality라고 \ \ 함) \\\\
\Longrightarrow \qquad f(E(X)) \le& E[f(X)] \quad(\because \ (\alpha) + (1 -  \alpha)=1, \ \  0  \le \alpha, \quad (즉 \ \ 확률분포라고 생각 \ \ 함)
\end{align}
$$

(iii) (ii)의 결과를 이용하면 다음과 같음
$$
\begin{align}
D_{KL}(P \| Q) =& \int_{\mathcal X} log_2 \left( \frac{P(x)}{Q(x)} \right) P(x) dx  \\
               =& \int_{\mathcal X} log_2 \left( - \frac{Q(x)}{P(x)} \right) P(x) dx \\
               \ge & \ \ log_2 \int_{\mathcal X} \left( - \frac{Q(x)}{P(x)} \right) P(x) dx \\
               =& \ \ - log_2 (1) \\
               =& \ \ 0
\end{align}
$$

$\quad$ 따라서 

$$
D_{KL}(P \| Q)\ge 0
$$

$\quad$ **특성 2**: KLD는 거리 개념이 아니고 asymmetric(비대칭)한 measure임

$$
D_{KL}(P \| Q) \neq D_{KL}(Q \| P)
$$

[증명]<br>
$D_{KL}(P \| Q) - D_{KL}(Q \| P)$를 계산해보자<br>

$$
\begin{align}
D_{KL}(P \| Q) - D_{KL}(Q \| P) =& H(p, q) -H(p) - \left[ H(q, p) -H(q) \right] \\
                                =& H(q) - H(p)  \qquad \because (H(p, q) = H(q, p)) \\
                                \neq & 0 
\end{align}
$$

$ \qquad  \triangle I_i = log_2(p_i) - log_2(q_i) $ <br>
$ \qquad  D_{KL}(P \| Q) = E \left[ log_2(p_i) - log_2(q_i) \right] 
               = \sum_i log_2 \left( \frac{p_i}{q_i} \right) p_i $

## 6. Jenson-Shannon Divergence
- KL-Divergence는 근사된 확률 분포가 얼마나 원본과 비슷한지를 측정하는 척도임<br>

- 더 엄밀하게 정의하면 
$D_{KL}( P\| Q)$는 확률분포 $P$를 확률분포 $Q$로 근사화할 때, 정보량의 차이(정보손실량)의 기댓값임<br><br>

- 그리고 KL-Divergence는 Symmetric $ D_{KL}(P \| Q) \neq D_{KL}(Q \| P) $하지 않음
 
- Jenson-Shannon Divergence는 두 분포간의 차이를 계산할 때 symmetric하게 사용할 수 있도록 변형한 measure로 거리의 개념을 가짐
 <br><br>
 
$$
JSD(P, Q) = \frac{1}{2} D_{KL} \left( P \| \frac{P+Q}{2} \right) 
           +
            \frac{1}{2} D_{KL} \left( Q \| \frac{P+Q}{2} \right) 
$$

## 7. Cross-Entropy

### 7.1 Cross-Entropy의 정의

- 인공지능에서 다루는 예측 모형은 참인 실제 분포인 $P(Y=y)$가 unknown이고, $P(Y=y)$로 부터 sampling된 data $y_1, ..., y_n$을 이용하여 <br> $ \  $ $P(Y=y)$ 분포를 예측하고자 하는 문제를 다룸<br><br>

- Sample을 통해 구해진  분포를 $Q(Y=y)$ 라고 할 때, 참인 실제 분포인 $P(Y=y)$와  추정한  $Q(Y=y)$에 대해, 추정한 $Q(Y=y)$가 $P(Y=y)$에 얼마나 잘 근사화 <br> $ \  $ 되었는지 판단하기 위해 다음과 같이 정의된 cross-entropy $H_p(q)$를 사용함

$$ 
\begin{align}
H_P(Q) =& E_P \left[ log \left( \frac{1}{Q} \right) \right]  \\\\
       =& 
\begin{cases} 
\sum_{y \in \mathcal{Y}} \left( log \frac{1}{Q(y)} \right) P(y),     &       if \ \ Y가 \ \ 이산형확률변수일 \ \ 때  \\\\
\int_{y \in Y} \left( log \frac{1}{Q(y)} \right) P(y)dy,   &       if \ \ Y가 \ \ 연속형확률변수일 \ \ 때
\end{cases}
\end{align}
$$ 

### 7.2 Cross-Entropy와 KLD의 관계
$$
\begin{align}
D_{KL}(P \| Q) =& E_P \left[ log \frac{P(Y)}{Q(Y)} \right] \\
               =& E_P \left[ log \frac{1}{Q(Y)} \right]
                +
                  E_P \left[ log {P(Y)} \right] \quad (\because \ Entropy의 \ 정의)\\
               =& H_P(Q) - H(P) \\\\
\Longrightarrow \qquad \quad & \\\\
H_P(Q) =& D_{KL}(P \| Q) + H(P)
\end{align}
$$

$\quad$ 즉, Cross Entropy $CE$는 $D_{KL}$에 상수 $H(P)$를 더한 값임<br> 
$\quad \ \ $ $\checkmark \ $ $CE$와 $D_{KL}$는 최적화 관점에서 동일한 measure임
