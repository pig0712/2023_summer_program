```python
df = pd.read_csv("./movies_metadata.csv")
df
```

# 1. 가장 런타임 긴 시간 영화 1편을 찾아주세요

```python
df["runtime"].idxmax()
```

# 2. 2016년 1월부터 3월까지 나왔던 시리즈 영화를 찾아주세요

```python
a = df[df['release_date'].between('2016-01-01', '2016-03-30')]
a['title']
```

# 3. 영화 제목이 "toy"로 시작하는 영화 개수를 찾아주세요

```python
aaa = df[df['title'].fillna("").str.startswith("Toy")]
len(aaa['title'])
```

# 4. 한국어 영화 중 평점 8.0이상인 영화를 찾아주세요

```python
a = df[df['original_language'] == 'ko']
b = a[a['vote_average'] >= 8]
b
```

# 5. 한국어 영화의 수익 평균을 찾아주세요

```python
a = df[df['original_language'] == 'ko']
a['revenue'].mean().round(2)
```
