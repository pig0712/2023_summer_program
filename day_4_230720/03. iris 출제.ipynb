# IRIS 데이터셋

- sepal_length : 꽃받침 길이
- sepal_width : 꽃받침 너비
- petal_length : 꽃잎 길이 
- petal_width : 꽃잎 너비
- species : iris 꽃 종류
    - setosa : 세토사
    - versicolor : 버시칼라
    - virginica : 버지니카
