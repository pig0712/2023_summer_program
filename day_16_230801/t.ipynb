```python
import cv2

src = cv2.imread("C:/Users/koll2/OneDrive/2023_summer_program/day_16_230801/Lenna.jpg", cv2.IMREAD_COLOR)
flipedX = cv2.flip(src, 0)
flipedY = cv2.flip(src, 1)
flipedXY = cv2.flip(src, -1)

cv2.imshow("src", src)
cv2.imshow("fliped with X", flipedX)
cv2.imshow("fliped with Y", flipedY)
cv2.imshow("fliped with XY", flipedXY)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2

src = cv2.imread("C:/Users/koll2/OneDrive/2023_summer_program/day_16_230801/Lenna.jpg", cv2.IMREAD_COLOR)

height, width, channel = src.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
rotatedImg = cv2.warpAffine(src, matrix, (width, height))

cv2.imshow("src", src)
cv2.imshow("rotated", rotatedImg)
cv2.waitKey()
cv2.destroyAllWindows()
```
