```python
data = [[5,25,50,20],[4,23,51,17],[6,22,52,19]]

X = np.arange(4)
plt.bar(X + 0.00, data[0], color = "r", width=0.25)
plt.bar(X + 0.25, data[1], color = "g", width=0.25)
plt.bar(X + 0.50, data[2], color = "b", width=0.25)
plt.show()
```

```python
color_list = ["r", "g", "b"]
gap = 0.8/ len(data)

for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i * gap, data[i], color=color_list[i], width=gap)
plt.show()
```

```python
A=[5,25,50,20]
B=[4,23,51,17]
C=[6,22,52,19]

X = np.arange(4)
plt.bar(X,A,color="r",width=0.8)
plt.bar(X,B,color="g",width=0.5, bottom=A)
plt.bar(X,C,color="b",width=0.25, bottom=B)

plt.show()
```

```python
data = np.array([[5,25,50,20],[4,23,51,17],[6,22,52,19]])
color_list = ["r","g","b"]

X = np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i], color_list[i], bottom=np.sum(data[:i], axis=0))
plt.show()
```
