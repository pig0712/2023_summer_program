```python
import pandas as pd

# 데이터 생성
data = [["익산", 123, 123], ["군산", 123, 123], ["전주", 123, 123]]

a = []
for i in range(3):
    a.append(input("값을 입력하세요: "))  # 사용자로부터 값을 입력받아 리스트에 추가

data.append(a)

# 데이터프레임 생성
df = pd.DataFrame(data, columns=["지역", "2023", "2022"])

# 파일 이름 설정
name = input("파일 이름을 입력하세요: ")


# 엑셀 파일로 저장
excel_path = f"C:/Users/koll2/Desktop/{name}.xlsx"
df.to_excel(excel_path, index=False)  # index를 제외하고 엑셀에 저장
```
