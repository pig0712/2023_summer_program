## 2. Series를 이용한 근사화

### 2.1 Taylor Series
- 미적분학(Differential and Integral Calculus)에서, 테일러 급수(Taylor Series)는 도함수(Derivative Function )들의 한 점에서의 값으로 계산된 항의 무한합으로 해석함수를 나타내는 방법<br>


- 복잡한 함수 $f(x)$를 우리가 잘 다룰 수 있는 $x^{1}, x^{2}, ..., $으로 변환하는 방법

##### (1) Taylor Series

#### $ \ \ $ [Taylor Series 정의] 
> Smooth Function $f: R \rightarrow R$ 및 실수(real number) $a \in R$이 주어졌을 때, $f$의 Taylor Series는 다음과 같은 Power Series이다 <br><br>
$$
\begin{align}
f(x) &=\sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x-a)^{n} \\
       &= f(a) + f^{(1)}(a)(x-a) + \frac{1}{2} f^{(2)}(a)(x-a)^2 + \frac{1}{6} f^{(3)}(a)(x-a)^3 + \cdots
\end{align}
$$


#### $ \ \ $ [Maclaurin Series] 
- $a=0$인 Taylor Series<br><br>
$$
\begin{align}
T_f(x) &=\sum_{n=0}^{\infty} \frac{f^{(n)}(0)}{n!}x^{n} \\
       &= f(0) + f^{(1)}(0)x + \frac{1}{2} f^{(2)}(0)x^2 + \frac{1}{6} f^{(3)}(0)x^3 + \cdots
\end{align}
$$


#### $ \ \ $ [다변수 Taylor Series] 

- Taylor Series는 2개 이상의 다변수로 일반화될 수 있음. $d$개의 변수를 가지는 smooth function $f: R^{d} \rightarrow R$의 Taylor는 다음과 같다.<br><br>
$$
\begin{align}
f(x_1, ..., x_d) &= \sum_{n_1=0}^{\infty} \cdots \sum_{n_d=0}^{\infty} 
     \frac{f^{(n_1)}(a_1) \cdots f^{(n_d)}(a_d)}{n_1! \cdots n_d!} (x_1-a_1)^{n_1} \cdots (x_d-a_d)^{n_d}\\
       &= f(a_1 \cdots a_d)  \\
       & \ \ \ + \frac{1}{1!} \sum_{j=1}^{d} \frac{\partial f(a_1, \cdots, a_d)}{\partial x_j} (x_j - a_j) \\
       & \ \ \ + \frac{1}{2!} \sum_{k=1}^{d} \sum_{j=1}^{d} 
                    \frac{\partial^2 f(a_1, \cdots, a_d)}{\partial x_j \partial x_k} 
                    (x_k - a_k)(x_j - a_j) \\
       & \ \ \ + \frac{1}{3!} \sum_{l=1}^{d} \sum_{k=1}^{d} \sum_{j=1}^{d} 
                    \frac{\partial^3 f(a_1, \cdots, a_d)}{\partial x_l \partial x_j \partial x_k} 
                    (x_l - a_l)(x_k - a_k)(x_j - a_j) + \cdots
\end{align}
$$

#### (2) Taylor Series를 이용한 선형 근사(Linear Approximation)
- 선형 근사(Linear Approximation)는 어떤 함수를 선형 함수, 즉 일차 함수 만으로 근사하는 것을 말함<br>
$ \ \ $ $\checkmark \ $ 어떤 점  $x=a$에서 미분가능한 함수 $f(x)$가 있을 때, $x=a$ 점에서 근사화한 함수는 다음과 같음
<br>
<br>
$$
f(x) = f(a) + f'(a)(x-a) + \varepsilon 
$$
<br>
$ \ \ $ $\checkmark \ $ 이 때 $f(x)$의 1차근사 approximation function은 다음과 같음
<br>
<br>
$$
f(x) \approx f(a) +f'(a)(x-a)$$
<br>
$ \ \ $ $\checkmark \ $ 위의 근사식을 $f$의 $a$에서의 선형근사(Linear Approximation)라고 한다


- 이것은 Tayler 정리에 의하여 얻어진 $
f(x) = f(a) + f'(a)(x-a) + \varepsilon 
$에서 approximation $\varepsilon \approx 0$를 취한 것으로 볼 수 있다. 

#### 연습문제 1
$\sqrt{4.01}$을 함수 $f(x)=\sqrt{x}$의 $a=4$에서의 Linear approximation을 사용해서 근사값을 구하시오

```python
import numpy as np

def f_sqrt(x):
    return np.sqrt(x)

def d1_sqrt(x):
    return 1/(2*np.sqrt(x))

def approx(x, a):
    return f_sqrt(a) + d1_sqrt(a)*(x-a)

approx(4.01, 4)
```

```python
np.sqrt(4.01)
```

```python
approx(4.01, 4) - np.sqrt(4.01)
```

```python
a_set = np.linspace(2, 6, 30)
print(a_set)
```

```python
app = approx(4.01, a_set)
error = np.abs(np.sqrt(4.01) - app)
error
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5, 3))
plt.plot(a_set, error)
plt.title('Linear Approximation Error')
plt.xlabel('a: Approximation Point')
plt.ylabel('Error')
plt.show()
```

### 2.2 Fourier series

- 푸리에 급수(Fourier級數, Fourier series)는 주기 함수를 삼각함수의 가중치로 분해한 급수임 <br>
$ \ \ $ $\checkmark $ 대부분의 경우, 급수의 계수는 본래 함수와 일대일로 대응함  <br>
$ \ \ $ $\checkmark $ 함수의 푸리에 계수는 본래 함수보다 다루기 쉽기 때문에 유용하게 사용됨  <br>


- 푸리에 급수는 전자 공학, 진동 해석, 음향학, 광학, 신호처리와 화상처리, 데이터 압축 등에 사용되고 있음


- 복잡한 함수 $f(x)$를 우리가 잘 다룰 수 있는 $sin  \theta, cos \theta, ..., $로 변환하는 방법

#### (1) Fourier Series 정의
> - 푸리에 급수는 주기함수를 기본적인 조화함수인 삼각함수 또는 복소 지수 함수의 급수로 나타낸 것임<br>
(1) 주기함수 $f: R \rightarrow C$(복소수)가 $T$의 주기를 가지는 함수라고 하자. 즉, 
<br>
<br>
$$
f(x) = f(x+T)
$$
<br>
(2) 또한 $f$가 모든 유한 구간(finite interval)에서 제곱적분 가능하다고 하자. <br> $\quad$ 즉, 임의의 $a, b \in R$에 대하여 $\int_{a}^{b} |f|^2 dx$가 유한한 값으로 존재한다고 하자<br>
그렇다면 $f$의 푸리에 계수(Fourier coefficient) $g_n$을 다음과 같이 정의한다<br><br>
$$
g_n = \frac{1}{T} \int_{t_0}^{t_0 + T} exp \left( - \frac{2n \pi ix}{T} \right) f(x) dx, t_o \in R
$$<br>
그렇다면 다음이 성립한다. 임의의 $x \in R$에 대하여, 다음식이 성립하지 않는 $x$의 집합은 로베그 측도* 0을 가진다
<br>
<br>
$$
\begin{align}
f(x) =& \sum_{n=-\infty}^{\infty} g_n  exp \left( - \frac{2n \pi ix}{T} \right)  \\
\rightarrow f(x) =& \sum_{n=-\infty}^{\infty} C_n \left( cos(n \omega t) + i sin(n \omega t) \right) \quad ( Euler's \ \ Rule)
\end{align}
$$
<br>
만약 $f$가 연속미분가능$\left( C^1 \right)$ 함수라면(즉, $f$의 도함수가 존재하고 연속적인 경우) $f$의 푸리에 급수는 모든 $x$에서 $f(x)$로 수렴한다


* 측도론에서 르베그 측도(영어: Lebesgue measure)는 유클리드 공간의 부분 집합에 길이, 넓이 또는 부피를 할당하는 방법

#### (2) Fourior Series를 이용한 계단함수 Approximation의 예

- 다음과 같은 계단함수를 Fourior Series로 Approximation하보면 다음과 같음

$$
\begin{align}
f(x) =&
\begin{cases} 
0, &       if \ 0 \le x  \lt 1   \\
1, &       if \ 1 \le x  \lt 2    \\
2, &       if \ 2 \le x  \lt 3
\end{cases} \\\\
f(x) =& 1 - \frac{3 sin \left( \frac{2 \pi x}{3} \right)}{\pi}
          - \frac{3 sin \left( \frac{4 \pi x}{3} \right)}{2 \pi}
          - \frac{3 sin \left( \frac{8 \pi x}{3} \right)}{4 \pi}
          - \frac{3 sin \left( \frac{10 \pi x}{3} \right)}{5 \pi}
          + \cdots
\end{align}
$$

### 3.3  Featue $z$와 Label $y$의 관계 파악: 공분산과 상관관계 

####  (1) 공분산(Covariance)

- 공분산(Covariance)은 <span style="color:red">두 확률변수(pairwise) 간</span>의 <span style="color:red">선형관계</span>를 파악하는 방법<br>
$ \ \ $ $\checkmark \ $ 동시에 여러개의 확률변수간의 관계를 이야기하는 것이 아님<br>
$ \ \ $ $\checkmark \ $ 확률변수 $X_1$와 $X_2$의 기댓값과 분산을 각각 $\left( \mu_{1}, \sigma_{1}^2 \right)$과   $\left( \mu_{2}, \sigma_{2}^2 \right)$라고 할 때 공분산$(Covariance)$은 다음과 같이 <br> $\quad \ \ $ 정의됨<br><br>

#### [Covariance 정의]
>$ \begin{align}
\qquad  Cov(X_1, X_2) & = E(X_1-\mu_1)(X_2-\mu_2)\\
                      & = E(X_1 X_2 - \mu_2 X_1 - \mu_1 X_2 + \mu_1 \mu_2)\\
                      & = E(X_1 X_2) - \mu_2 E(X_1) - \mu_1 E(X_2) + \mu_1 \mu_2\\
                      & = E(X_1 X_2) - \mu_1 \mu_2
\end{align}$<br><br>

#### [공분산행렬(Covariance Matrix)]

- Column vector인  확률변수 $X=(X_1, ..., X_m)$와 $Y=(Y_1, ..., Y_n)$의 기댓값이 각각 $\mu_X$와 $\mu_Y$라고 할 때 $m \times n$ 공분산행렬은 다음과 같다<br>

$$
\begin{align}
Cov(X, Y) =& E\left((X-\mu_X)(Y-\mu_Y)^T\right) \\
Cov(X, Y) =& Cov(Y, X)^T
\end{align}
$$

#### [표본공분산(Sample Covariance)]

- 피어슨 상관계수에 사용되는 표본 공분산(Sample Covariance)은 다음과 같다.<br>

$$
\widehat{Cov}(X, Y) = \frac{\sum_{i=1}^{n}(X_i- \bar X)(Y_i- \bar Y)}{n-1}
$$

####  (2) 상관관계(Correlation Coefficient)

- 상관관계는 2개의 Feature간에 선형관계 정도를 Measure하는 방법임<br>
$ \ \ $ $\checkmark \ $ 비선형 관계는 파악하지 못함<br>
$ \ \ $ $\checkmark \ $ 2개 이상의 Feature가 관련성을 가질 때 파악할 수 없음 <br>


- 모집단(Population)에 적용되는 Pearson's correlation coefficient, $\rho_{X_1 X_2}$<br>
<br>

$$ 
\begin{align}
\rho_{X_1 X_2} &= \frac{Cov(X_1, X_2)}{\sigma_{X_1}\sigma_{X_2}}
\end{align}
$$<br>

$$ -1\le \rho \le 1$$
<br>
- 표본(Sample)에 적용되는 Pearson's correlation coefficient, $r_{X_1 X_2}$<br>


$$ 
\begin{align}
r_{X_1 X_2} = \frac{S_{X_1 X_2}}   {S_{X_1 X_1} S_{X_2 X_2}}
\end{align}
$$<br>
<br>
$ \qquad$ 여기서 $S_{X_1 X_2} = \sum_{i=1}^n (x_{1i}-\bar{x_1})(x_{2i}-\bar{x_2})$, 
$S_{X_1 X_1} = \sqrt{     \sum_{i=1}^n (x_{1i}-\bar{x_1})^2   }$,

### 3.4  Featue $z$와 Label $y$의 관계 파악: 선형회귀분석
#### (1) 선형회귀의 Model과 가정
- 선형회귀(Linear regression)는 종속 변수 $y$와 한 개 이상의 Feature (독립 변수 또는 설명 변수) $z=\left(z_1, \cdots ,z_{n_0}  \right)$와의 선형 관계를 추정하는 통계적 모델링 기법


$$ 
\begin{align}
Model에 \ 대한 \ 가정: \ y &= w_0 + w_1 z_1 + \cdots + w_{n_0} z_{n_0} + \varepsilon \quad (선형근사) \\\\
Error에 \ 대한 \ 가정: \  \varepsilon & \sim (0, \Sigma), \quad Constant \ variance
\end{align}
$$ <br>

$$
\begin{align}
\Sigma
 &=
\begin{bmatrix}
\sigma_{11} & \cdots &  \sigma_{1c} \\
\vdots & \ddots & \vdots \\
\sigma_{c1} & \cdots &  \sigma_{nn}
\end{bmatrix} \\\\
\sigma_{ij} &= E \left[ \varepsilon_i - E \left( \varepsilon_i \right) \right] \left[ \varepsilon_i - E \left( \varepsilon_i \right) \right] = E \left(  \varepsilon_i \varepsilon_j \right)
\end{align}
$$
