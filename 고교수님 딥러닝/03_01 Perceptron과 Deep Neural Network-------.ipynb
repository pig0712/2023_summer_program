## 3. Perceptron Model

- Perceptron은 <br>
$ \ \ $ $\checkmark \ $ perception과 neuron의 합성어로 <br>
$ \ \ $ $\checkmark \ $ 생물의 신경계(neual network)를 구성하는 기본 단위인 neuron의 동작을 modeling한 algorithm임<br>


- Perceptron은 다수의 input을 입력받아 하나의 값을 출력하는 algorithm임 <br>
$ \ \ $ $\checkmark \ $ 단층 퍼셉트론: 초기 𝑁𝑒𝑢𝑟𝑎𝑙 𝑁𝑒𝑡𝑤𝑜𝑟𝑘(1957, 𝐹𝑟𝑎𝑛𝑘 𝑅𝑜𝑠𝑒𝑛𝑏𝑙𝑎𝑡𝑡 개발) <br>
$ \ \ $ $\checkmark \ $ 입력 feature들의 선형결합 값을 구하고 그 값이 0보다 큰지를 판단

$$
\begin{align}
-0.5 + z_1 - z_2 > 0 \ \rightarrow &  \ 양품  \\
-0.5 + z_1 - z_2 \le 0 \ \rightarrow &  \ 불량품
\end{align}
$$

### 3.1 Perceptron을 이용한 부호 변환기

- 부호 변환기는 <br>
$ \ \ $ $ \checkmark \ $ $True$가 input되면 $False$를 출력하고  <br>
$ \ \ $ $ \checkmark \ $ $False$가 input되면 $True$를 출력하는 수식 또는 Network을 말함<br>


- 부호변환기 수식 표현

$\qquad \qquad$  (1) $ y \ = \ f(z) \ = \ w_0 + w_1 z_1 =  \ 0.5 - z_1 $ ($w_0=0.5, w_1=-1$은 주어진 값임)   <br>

$\qquad \qquad$ (2) $ \ \ $  $y$값이 $y \le 0$ 이면 $ y \ \leftarrow \ 0 $<br>
$\qquad \qquad \quad \ \ \ $ $y$값이 $y \gt 0$ 이면 $ y \ \leftarrow \ 1 $
<br>

$ \qquad $ $\checkmark \ $ (부호 변환기 적용) 

$\qquad \qquad$ $z_1= 1$이면 <br> 
$\qquad \qquad \quad$ $\cdot$ $ y = 0.5 - 1 = -0.5$ <br>
$\qquad \qquad \quad$ $\cdot$ $ y \le 0 $이므로 $ y=0 $<br>


$\qquad \qquad$ $z_1= 0$이면 <br> 
$\qquad \qquad \quad$ $\cdot$ $ y = 0.5 + 0 = +0.5$ <br> 
$\qquad \qquad \quad$ $\cdot$ $ y \ge 0$이므로 $ y=1$
$$

- 부호변환기 Network 표현<br>
$ \ \ $ $\checkmark \ $ 하나의 점으로 Calssification할 수 있음

#### [Classification]의 의미

>-  Classification이라고 함은 Data $z \in R^{n_{_0}}$보다 1차원 이상 낮은 Hyperplane $H \in R^{k}$ Data $z_i$들을 구분하는 것을 말함<br>
>-  단 $k \le \left( n_{_0}-1 \right)$

```python
import numpy as np

weight = np.array([0.5,  -1.0])

def Decision(x, weight):
    y = weight[0] + weight[1]*x
    if y > 0: 
        return 1
    else:
        return 0
```

```python
Decision(0, weight), Decision(1, weight)
```

### 3.4 Perceptron 한계극복을 위한 Nonlinear Transformation의 도입

- 선형Model로 구분할 수 없는 Classification문제인 경우, 다음과 같은 방법으로 Classification을 수행할 수 있음<br>
$ \ \ $ (i) 2개 이상의 Linear Hyperplane으로 Classification을 수행<br>
$\quad$ $\cdots \ $ 많은 수의 Linear Hyperplane을 사용하면 어떤 문제도 Classification할 수 있음<br>
$ \ \ $ (ii) 1개의 Linear Hyperplane과 매우 간단한 비선형 함수를 이용하여 Classification<br>


$\quad$ $\Longrightarrow \ $ <span style="color:red">2개 이상의 Linear Hperplane을 사용하는 것 보다 더 작은 수의 parameter를 가지는 (ii) 번 방법을 활용함</span>


- Activation Function이라고 부르는 선형변환 후, 비선형함수 $\sigma$를 도입하여 더 작은 $Loss$값을 가지는  Classification을 달성할 수 있음

## 4. Multi-Layer Perceptron Model
### 4.1 Perceptron 개념의 변화

- Multi-Layer Perceptron Model은 input을 (선형)변환하는 층(layer)를 2개 이상 가지는 Neural Network Model임<br>


- 각 Layer에서 Layer input에 해당하는 vector를 (i) 선형변환하는 과정과 (ii) 선형변환 결과를 비선형 변환하는 과정을 둠<br>
$ \ \ $ $\checkmark \ $ 비선형함수를 이용한 계산 단계에 사용하는 비선형 함수를 Activation Function이라고 부름<br>
$ \quad $ $\cdots \ $ 자주 사용하는 Activation Function으로는 𝑆𝑖𝑔𝑚𝑜𝑖𝑑, 𝑅𝑒𝐿𝑈, 𝑡𝑎𝑛ℎ𝑥 등이 있음<br>


- 비선형 활성함수를 사용하는 다층 퍼셉트론을 인공신경망 $(Artificial \ Neural \ Network)$ 또는 <br> $ \ \ $ 심층신경망$(Deep \ Neural \ Network)$이라고 부름

### 4.2 Multi-layer Perceptron 

#### (1) Multi-layer Perceptron의 Architecture

- 선형변환과 비선형변환으로 구성되는 layer를 반복한 구조


- Multi-layer Perceptron Model부터는 모든 layer에서 각 선형변환된 node $y_j^{(l)}$에 bias node $w_{0,j}^{(l)}$을 생략하고 표현하지 않음 

#### (2) 각 Layer 설명 

- 입력층(𝐼𝑛𝑝𝑢𝑡 𝑙𝑎𝑦𝑒𝑟) <br>
$ \ \ $ $\checkmark \ $ $n_{_0} \ := \ $ Input Feature의 수(또는 차원)<br>
$ \ \ $ $\checkmark \ $ $z^{(1)} := \left( z_1^{(1)}, z_2^{(1)}, ..., z_{n_{_0}}^{(1)} \right)^T$ input feature 변수(vector)<br>
<br>
- 은닉층(𝐻𝑖𝑑𝑑𝑒𝑛  𝑙𝑎𝑦𝑒𝑟)<br>
$ \ \ $ $\checkmark \ $ $L$$ \ := \ $ 은닉층수(hyperparameter임)<br>
$ \ \ $ $\checkmark \ $ $n_l \ := \ $ $l^{th}$ 은닉층에 포함된  선형변환된 node의 수(hyperparameter임) <br>
$ \ \ $ $\checkmark \ $ $y^{(l)} := \left( y_1^{(l)}, y_2^{(l)}, ..., y_{n_l}^{(l)} \right)^T = W_{n_l \times (n_l-1)}^{(l)} z_{(n_l-1) \times 1}^{(l)}$, $l^{th}$ layer의 input인 $z^{(l)}$을 $W^{(l)}$행렬을 이용하여 <br> $\qquad \qquad$ 선형변환한 결과 vector<br>
$ \ \ $ $\checkmark \ $ $w_{ij}^{(l)} \ := \ $ $l^{th}$ layer에서 node $z_i^{(l)}$에서 node $y_j^{(l)}$로 연결하는 parameter(가중치)<br>
$ \ \ $ $\checkmark \ $ $z^{(l+1)} \ := \ \sigma^{(l)} y^{(l)}$, $y_i^{(l)}$을 비선형 활성함수인 activation function $\sigma^{(l)}$을 Conponentwise로  적용하여 <br> $\qquad \qquad$  얻어진 결과 vector <br>
<br>
- 출력층(𝑂𝑢𝑡𝑝𝑢𝑡 𝑙𝑎𝑦𝑒𝑟)<br>
$ \ \ $ $\checkmark \ $ [범주형] 출력  node수$ \ := \ $  출력변수 class 수<br>
$ \ \ $ $\checkmark \ $ [연속형] 출력 node수$ \ := \ $출력 변수의 수<br>
<br>

- Multi-layer Perceptron의 Hyperparameter<br> 
$ \ \ $ $\checkmark \ $ Deep Neural Network Architecture를 결정하기 위해 인공지능 분석자가 정해 주어야하는 상수 값<br>
$ \ \ $ $\checkmark \ $ 각 hidden layer에서 선형변환에 사용되는 행렬 $W^{(l)}$의 차원이 $ \left( n_{n-1} \times n_l \right)$라는 의미는 각 hidden layer를  <br> $\quad \ \ $ 통과하며 차원변화가 발생한다는 의미임

#### (3) <span style="color:red">Deep Neural Network의 Vector$-$Matrix 표현</span>

$$
\begin{align}
& w_{i,j}^{(l)} \ := \ l^{th} hidden \ layer에서 \ node \ i에서 \ node \ j로 \ 연결하는 \ parameter  \\\\
& w_{.,j}^{(l)} \ := \begin{bmatrix}
w_{0,j}^{(l)}     \\
w_{1,j}^{(l)}     \\
w_{2,j}^{(l)}     \\
\vdots     \\
w_{n_l - 1,j}^{(l)}      
\end{bmatrix}  \ \   l^{th} hidden \ layer에서 \ 임의의 \ node에서 \ node \ j로 \ 연결하는 \ parameter(w_{0,j}^{(l)} \ := \ bias 포함)  \\\\
& W^{(l)} \ := \quad   \left(w_{.,1}^{(l)}, \qquad w_{.,2}^{(l)}, \quad ..., \quad w_{.,j}^{(l)}, \quad ..., \quad w_{.,n_l}^{(l)}  \right) \\\\
& \qquad \ = \ \
\begin{bmatrix}
w_{0,1}^{(l)}     & w_{0,2}^{(l)}        & ...     & w_{0,j}^{(l)} & ...   & w_{0,n_l}^{(l)}     \\
w_{1,1}^{(l)}     & w_{1,2}^{(l)}        & ...     & w_{1,j}^{(l)} & ...   & w_{1,n_l}^{(l)}     \\
w_{2,1}^{(l)}     & w_{2,2}^{(l)}        & ...     & w_{2,j}^{(l)} & ...    & w_{2,n_l}^{(l)}      \\
\vdots            & \vdots               & \ddots  & \vdots  & \ddots       & \vdots               \\
w_{n_l-1,1}^{(l)} & w_{n_l-1,2}^{(l)}    & ...     & w_{n_l-1,j}^{(l)} & ...  & w_{n_l-1,n_l}^{(l)}
\end{bmatrix}  \ \   l^{th} hidden \ layer에 \ 포함된 \ 모든 \ parameter  \\\\
& W \ := \  \left( W^{(1)}, W^{(2)}, ..., W^{(L)}  \right) \ \  Deep \ Neural \ Network에 \ 포함된 \ 모든 \ parameter  \\\\
& y_{j}^{(l)} \ = \ w_{.,j}^{(l)}z^{(l)} + \varepsilon_{j}^{(l)} \\\\
& \underset{n_{_l} \times 1}{ y^{(l)} } \ = \ \underset{n_{_l} \times n_{_{l-1}}}W^{(l)} \underset{n_{l-1} \times 1}{ z^{(l)} } + \underset{n_{_l} \times 1}{ \varepsilon^{(l)} }
\end{align}
$$

- Deep Neural Network으로 Approximation한다는 의미<br>
$ \ \ $ $\checkmark \ $ 모집단으로 부터의 Sample인 Input Data(vector) $z^{(1)}$에 대해 다음과 같은 함수를 사용하여 Approximation하는 것임  <br> 
$ \quad \ \ $ (i) 많은 수의 선형변환과 많은 수의 비선형변환 적용  <br> 
$ \qquad \ \  $ $\cdots \ $ 선형변환을 위한 $ \left( {n_{_{l-1}} \times n_{_l}} \right) $인 Matrix  $W^{(l)}$을 곱해줌에 따라 vector의 차원변화가 발생함 <br>
$ \quad \ \ $ (ii) (i)의 반복  
<span style="color:red">
$$
y = \sigma^{(L)} W^{(L)} \cdots \sigma^{(2)} W^{(2)} \sigma^{(1)} W^{(1)}  \left( z^{(1)} \right) + \varepsilon
$$
</span>
