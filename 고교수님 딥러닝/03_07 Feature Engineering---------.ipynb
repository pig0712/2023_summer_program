## 1. Feature란 무엇인가?

### 1.1 Feature

#### (1) Feature의 개념

>1. Feature란 <span style="color:red">관찰되는 현상</span>의 <span style="color:red">측정 가능</span>한 특징임<br>
$\quad$ (i) Label을 $y$라고 할때, Feature $z = \left( z_1, z_2, ..., z_{n_{_0}} \right)$는 Machine Learning에서 입력이자  <br> $\qquad$ 설명 변수이며 때로는 함수 형태임<br>
><br>
>$$
y = f \left( z_1, z_2, ..., z_{n_{_0}} \right)
$$
><br>
>$\quad$ (ii) Feature는 때때로 속성(attribute)과 동일시되지만, Feature는 다른 Feature의 함수이거나 <br> $\qquad$ 특정 Feature의 또 다른 표현일 수 있음<br><br>
>
>2. Feature는 Machine Learning Model에서 <span style="color:red"> 예측값을 계산</span>할 수 있도록 숫자 값(vector)(또는 부울값)으로 표현되어야 함<br>
$\quad$ (i) Machine Learning에서 label $y$에 대한 예측치는 숫자(또는 vector)이지만 학습기에 따라 규칙과  <br> $\qquad$  가중치가 적용될 수 있기 때문에 featrure $z$는 숫자(또는 vector) 또는 boolean이어야함<br>
$\quad$ (ii) 즉, Feature는 Machine Learning 과정에서 선형변환함수 및 비선형변환함수를 통해 다양하게  <br> $\qquad$  변환됨<br><br>
>
>3. 어떤 Machine Learning Project는 성공하고 또 어떤 project는 실패함. 성공과 실패라는 차이는 왜 생길까?<br>
$\quad$ (i) 간단히 설명하면 성공, 실패를 좌우하는 가장 중요한 요소는 Machine Learning에 사용된 Feature임 <br>
$\quad$ (ii) 어떤 원시 Data는 학습할 수 있는 형식이 아닐 수 있지만, 해당 원시 Data를 이용하여 <br> $\qquad$ Feature들을 구성할 수 있음<br>
$\quad$ (iii) Machine Learning Project에서 이러한 Feature를 구성하는 과정에 대부분의 시간과 노력이 투입됨<br><br>
>
>4. 좋은 Feature를 구성하기 위해서는 전문지식, 기술적인 요소 뿐 아니라 직관, 창의성 및 종합적 사고가 필요함<br>

#### (2) Feature(특성)의 정의

><span style="color:red"> **[정의] Feature**  <br>
각 관찰 data(raw data)에서 발견되는 고유한 성질 및 특징 등을 의미 </span> <br>

$$ \underbrace{y}_{Label} = f \underbrace{\left( z_1, ..., z_{n_{_0}} \right)}_{Features} $$

$\qquad$ 즉, feature $  z = \left( z_1, ..., z_{n_{_0}} \right)  $는 <br> 
$\qquad \quad$ (i) $Label \ y$에 영향을 미치는 변수 <br> 
$\qquad \quad$ (ii) 또는 고유한 성질이라고 생각되는 $z$를 의미함 <br>
$\qquad \quad$ (iii) 많은 경우 label $y$에 대한 지식 또는 정보 없이 $z = \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right)$ 만이 관측됨

#### $\quad$ [통계학의 독립변수]  

$\qquad$ (1) 통계적 선형회귀 Model에 나타나는 Feature $z = \left( z_1, ..., z_{n_{_0}} \right)$ 를 생각해보자 <br>

$\qquad \qquad$ $(Model)$ <br>

$$
\begin{align}
y_i =& f \left( z_{i,1}, z_{i,2}, ..., z_{i, n_{_0}} \right)  \\
  =& w_0 + w_1z_{i,1} + w_2z_{i,2} + ... + w_{n_{_0}}z_{i, n_{_0}} +  \varepsilon_i, \  i=1,..., n  
\end{align}
$$

$\qquad \qquad$ $(Assumptions)$ 

$
\qquad \quad \qquad  (i) \ \ (zero \  mean \ and \ constant \ variance) \ \ \varepsilon_i \sim  (0, \sigma^2)이고  \\
\qquad \quad \qquad  (ii) \ (Statistically \ independent)\qquad \quad p \left( \varepsilon_i, \varepsilon_j \right) = p \left( \varepsilon_i \right) p \left( \varepsilon_j \right)
$<br><br>

$\qquad$ (2) 선형회귀분석에서의 $z=\left( z_1, ..., z_{n_{_0}} \right)$의 명칭 <br>

$\qquad \quad$ (i) 선형회귀분석(통계학)에서 $z$를 독립변수(independent variable), 예측변수(predictor variable),  <br> $\qquad \qquad$ 통제변수(controlled variable),   조작변수(manipulated variable) 등 다양한 이름으로 불리고 있음 <br>  
$\qquad \quad$ (ii) 기계학습(Machine Learning) 혹은 패턴인식(Pattern Recognition)에서 independent variable이라는 <br> $\qquad \qquad$ 용어 대신 Feature라고 부름<br> 

$\qquad \quad$ (iii) Machine Learning에서, 통계학에서 사용하는 independent variable이라는 용어를 피하는 이유는 <br> $\qquad \qquad \  $ 통계학에서 independent variable이   만족해야하는  성질을 충족하지 못할 수 있으므로 용어를 <br> $\qquad \qquad \  $구별하는 것임 <br><br>

$\qquad \qquad$ $\bullet \ $ 통계학에서 독립변수들의 각 조건 $z_i = \left( z_{i,1}, z_{i,2}, \cdots, z_{i, {n_{_0}}} \right)$에서 관측치 $y_i$가 얻어진 경우 <br> $\qquad \qquad \quad  $ dataset $(Z, y)$는 다음과 같이 표현됨

$\qquad \qquad \quad$ $\checkmark \ $ 통계학에서 일반적인 가정 중의 하나는 $Z$를 구성하는 열벡터 $z_j$들이 서로 독립(두 벡터의 <br> $\qquad \qquad \qquad  $ 내적이 0)이라고 가정함

$$
z_j \cdot z_k = 0, \ \ i \ne k
$$

$\qquad \qquad \qquad \ \ $ 여기서 $z= \left[ z_1, z_2, ..., z_{n_{_0}} \right]^T$

#### Feature의 예

- 어린이의 키 $y$에 영향을 미칠 것으로 예상되는 변수$\left(feature  \ z= \left( z_1, ..., z_{n_{_0}} \right) \right)$가 어떤 변수들로 구성되어 있을지 생각해 보자

 구분 | 변수명 | 변수 설명 | 변수로 선정한 이유
:----:|:----:|:----:|:----
<br> Label | <br> $y$ | <br> 어린이의 키 | <br> $\bullet \ $ 관심대상
<br> Features | <br> $z_1$ | 유전자 | <br> $\bullet \ $ 키가 큰 부모의 유전자를 가진 어린이는 키가 크다
        | <br> $z_2$ | <br> 영양섭취량 | <br> $\bullet \ $ 단위시간당 영양섭취량이 높은 어린이의 키가 크다
        | <br> $$ \vdots $$| <br> $$ \vdots $$ | <br> $$ \vdots $$
        | <br> $z_{n_{_0}}$ | <br> 수면시간 | <br> $\bullet \ $ 충분한 수면은 키성장에 도움이 된다

- 자주 만날 수 있는 다양한 Feature들의 예 <br>
$\quad$ $\checkmark \ $ Computer Vision에서 image는 관찰값이지만  image에 포함된 선은 Feature 일 수 있음<br>
$\quad$ $\checkmark \ $ 자연어 처리에서는 문서나 트윗이 관찰값이 될 수 있고 구 또는 단어 수가 Feature가 될 수 있음<br>
$\quad$ $\checkmark \ $ 음성 인식에서 발화는 관찰값일 수 있지만 한 단어는 Feature 일 수 있음

$\quad$ (ii) 비지도학습에서는 다양한 type의 data(feature)의  유사도 또는 거리를 이용하여 같은 유형의 feature끼리 <br> $\qquad$ grouping하게 됨<br><br>

### 1.2 다양한 Feature 선택의 예

#### (1) 날짜$-$시간 data의 분해

- 날짜$-$시간에는 Model로 쉽게 처리할 수 없는 많은 정보가 포함되어 있음<br>
$ \ \ $ $\checkmark \ $ 예: 2014–09–20 20:45:40


- 예를 들어, 시간은 다른 속성과 관계가 있다고 합리적으로 생각할 수 있음<br>
$ \ \ $ $\checkmark \ $ 매일 체중이 늘어나는 병아리의 몸무게를 시간에 따라 측정했지만, 여기서 말하는 시간은 병아리가 <br> $\quad \ \ $ 섭취한 사료의 양을 대변하는 것일 수 있음<br>
$\quad $ $\cdots \ $ 이때 우리는 시간이 경과함에 따라 병아리의 무게가 증가한다고 말하기도 하지만 병아리의 무게에 진실로 <br> $\qquad \ \ $ 영향을 미치는 변수는 병아리의 사료 섭취량임

$$
y =
\begin{cases} 
f(t), &       t는 \ 시간 \\\\
g(x), &       x는 \ 사료 \ 섭취량 
\end{cases}
$$


- 시간 $t$에 대한 회귀 Model에서, 시간 $t$ 대신에 "Hour of Day"라는 새로운 숫자 Feature를  만들 수 있음<br>
$ \ \ $ $\checkmark \ $ 시간 $t$대신에 오전, 정오, 오후, 밤 이렇게  4개 값을 가지는 새로운 서수 Feature(Ordinal Feature)을 만들 수 있음<br>
$ \ \ $ $\checkmark \ $ 이러한 Ordinal Feature는 의사 결정 트리 Model에 유용할 수 있음<br>
$ \ \ $ $\checkmark \ $ 유사한 접근 방식을 사용하여 1년 동안의 시간 관계, 월 관계 및 계절성의 다양한 구조를 선택할 수 있음

#### (2) 백분율(Percentages)

- 비율이나 백분율과 같은 연속적인 숫자 속성을 다룰 때 종종 정밀도가 높은 원시 값이 필요하지 않을 수 있음


- 따라서 이러한 고정밀 백분율을 숫자 정수로 반올림하는 것이 종종 합리적임


- 그런 다음 이러한 정수를 원시 값 또는 범주형(이산 클래스 기반) 기능으로 직접 사용할 수 있음

ID | Percent | 10 Scale | 100 Scale
:----:|:----:|:----:|:----:
id_01 | 0.9826 | 10 | 98 
id_02 | 0.6723 | 07 | 67 
id_03 | 0.5213 | 05 | 52 

#### (3) 범주형 속성분해(Decompose Categorical Attributes)

- 3가지 색구분인 Red, Blue 또는 Unknown가  "Item Color"라는 feature의 범주 속성이라고 하자


- Unknown color는 특별할 수 있지만, 수학적 model 또는 computer에서는 또 하나의 color처람 처리함<br>


- 정보 없음이라는 unknown이 가지는 정보를 보다 더 잘 노출하는 것이 도움이 될 수 있음


- "색상 있음"이라는 새 이진 기능을 만들고 항목에 색상이 있는 경우 값 "1"을 할당하고 색상을 알 수 없는 경우 "0"을 할당할 수 있음


- 한 단계 더 나아가 항목 색상이 가진 각 값에 대해 이진 기능을 만들 수 있음<br>
$ \ \ $ $\checkmark \ $ Is Red, Is Blue 및 Is Unknown의 세 가지 이진 속성임


- 이러한 추가 기능은 항목 색상 기능 대신(간단한 선형 Model을 시도하려는 경우) 또는 추가로(의사결정 트리와 같은 것을 최대한 활용하려는 경우) 사용할 수 있음

#### (4) Binning

- 원시 연속 숫자로 작업할 때 발생하는 문제는 이러한 연속숫자  분포가 종종 왜곡된다는 것임<br>
$ \ \ $ $\checkmark \ $ 예를들면  일부 값은 매우 자주 발생하지만 일부는 매우 드물다는 것임


- 이 외에도 이러한 연속숫자 값의 범위가 다양하다는 또 다른 문제가 있음<br>
$ \ \ $ $\checkmark \ $ 예를 들어 특정 뮤직 비디오의 조회수가 비정상적으로 클 수 있으며 일부는 작을 수 있음


- 양자화(quantization)라고 하는 비닝(binning)은 연속 숫자인 feature를 이산 기능(범주)으로 변환하는 데 사용됨<br>
$ \ \ $ $\checkmark \ $ 이러한 불연속 값 또는 숫자는 원시 연속 숫자 값이 구간화되거나 그룹화되는 범주 또는 구간화로 생각할 수 있음<br>
$ \ \ $ $\checkmark \ $ 각 bin은 특정 강도의 크기를 나타냄

#### (5) Grouping

- 대부분의 Dataset에서 모든 Data들은 Dataset의 행(row)으로 표시되며, 모든 열은 Feature를 의미함<br> 
$ \ \ $ $\checkmark \ $ 이러한 종류의 Data를 "잘 정돈된(Tidy)"이라고 말함 


- 거래 data(transactions data)과 같은 Dataset은 인스턴스의 여러 행으로 인해 위의 깔끔한 Data 정의에 거의 맞지 않음<br> 
$ \ \ $ $\checkmark \ $ 이러한 경우 Data를 인스턴스별로 Group한 다음 모든 인스턴스를 하나의 행으로만 표시함<br>
$ \ \ $ $\checkmark \ $ Dataset 조작 방법 중 하나인 groupby의 핵심은 기능의 집계 기능을 결정하는 것임


- Numerical Feature의 경우 평균 및 합계 기능이 일반적으로 편리한 옵션인 반면 Categorical Feature인 경우 훨씬 더 복잡함<br> 
$ \ \ $ $\checkmark \ $ 숫자 열은 대부분의 경우 합계 및 평균 함수를 사용하여 그룹화됨 <br>
$ \ \ $ $\checkmark \ $ Feature의 의미에 따라 둘 다 바람직할 수 있음<br>
$ \ \ $ $\checkmark \ $ 예를 들어 비율 열을 얻으려면 이진 열의 평균을 사용할 수 있음


- 숫자가 아닌 feature의 경우 첫 번째 옵션은 빈도가 가장 높은 label을 선택하는 것임

### 1.3 특성(Feature) 선택

#### (1) Feature $ \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right) $가 얻어지는 상황의 이해

#### (1.1) Unknown 세상(Population) 

- $\left( x_{1}, x_{2}, ..., x_{k} \right )$는 Label $y$에 영향을 미치는 참인 Feature들이고  Label $ㅛ$와 Feature $\left( x_{1}, x_{2}, ..., x_{k} \right )$ 사이에 관계를 다음과 같이 표시하자 

$$
y = f \left( x_{1}, x_{2}, ..., x_{k} \right)
$$

$\qquad$ 여기서, <br>
$\qquad\quad$ (1) Label $y$에 영향을 미치는 참인 Feature $\left( x_{1}, x_{2}, ..., x_{k} \right)$를 알 수 없음<br>
$\qquad\quad$ (2) Label $y$와 참인 Feature $\left( x_{1}, x_{2}, ..., x_{k} \right)$ 사이의 함수관계 $f$에 대해서도 모름<br>
$\qquad\quad$ (3) 단지  Sampling을 통해 $n$개의 Sample은 $ y_{1}, y_{2}, ..., y_{n} $ 관측 가능함

#### (1.2) Feature로  $ \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right) $을 도입한  의미

- 많은 인공지능 문제에서 True Feature  $ \left( x_{1}, x_{2}, ..., x_{k} \right) $는 알수 없음


- 우리가 Machine Learning에서 사용하는 Feature는 Machine Learning Engineer 및 전문가가 보유하고 있는 지식을 이용하여 선택된 Feature  $ z = \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right) $임


- 관측치(또는 Label) $y_i$를 관측할 때($y$ Data를 Sampling 할 때), Machine Learning Engineer 및 전문가가 선택한  <br>  Feature들 $ \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right) $는 Machine Learning Engineer 및 전문가의 사전지식을 기반으로 도입한 변수임<br>
$ \ \ $ $\checkmark \ $ $y: \ $ 어린이의 키<br>
$ \ \ $ $\checkmark \ $ $z_1: \ $ 유전자<br>
$ \ \ $ $\checkmark \ $ $z_2: \ $ 1일 영양섭취량<br>
$ \ \ $ $\checkmark \ $ $z_3: \ $ 수면시간<br>




항목  |   $\qquad \qquad \qquad$ 설명
:----:|:----
<br> $\mathcal{X}$ | <br> $\bullet \ $ True feature set <br> $ \ \ $ $\checkmark \ $ True feature $ x_i \in \mathcal{X}$, $\forall \ i$  
<br> $\mathcal{Z}$ | <br> $\bullet \ $ 선택된 feature set <br> $ \ \ $ $\checkmark \ $ 사전지식으로 선택된 feature $ z_i \in \mathcal{Z}$, $\forall \ i$ 
<br> 관측치 또는 <br> Label $y$ |  <br> $\bullet \ $ $f$는 알지 못하는 함수 <br> $\bullet \ $ $y$에 영향을 미치는 $x = \left( x_1, x_2, ..., x_k \right)$도 알지 못함 <br> $\bullet \ $  $f$와 $x$는 모르지만 sample $y_i, \ i=1, ..., n $은 획득할 수 있음 <br>
<br>Feature들 |    <br>$\bullet \ $ 우리의 <span style="color:red">사전지식</span>을 이용하여 <span style="color:red"> $y$에 영향을 미칠 것이라고 예상되는   변수(feature)를 도입함 </span> <br><br> $$ z = \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right) $$ <br>  $\bullet \ $ 우리가 도입한 feature $ \left( z_{1}, z_{2}, ..., z_{n_{_0}} \right) $들은  다음과 같은 경우로 구분할 수 있음 <br> $\quad$ (Case 1)  $ z_i \in \mathcal{X}$  $\Longrightarrow$ feature $z_i$를 잘 선택한 경우 <br>  $\quad$ (Case 2)  $ z_i \notin \mathcal{X}$  $\Longrightarrow$ $y$에 영향을 미치지 않는 의미없는 feature $z_i$를  선택한 경우 <br> $\qquad \quad $ $ \checkmark \ $ 불필요한 $z_i$를 제거하기 위한 방법이 필요 <br>  <br>  $\bullet \ $ 또한 Unknown Feature $\mathcal{X}$와 우리가 선택한 feature $ \mathcal{Z}$ 사이에  다음과 같은 경우가 존재함<br><br>   $\quad$ (Case 1) $\mathcal{X} = \mathcal{Z}$인 경우   <br> $\qquad $ $\Longrightarrow$ 우리의 사전지식으로 $y$를 설명할 수 있는 True eature를 선택한 경우 <br><br>   $\quad$ (Case 2) $\mathcal{X} \subset \mathcal{Z}$인 경우 <br> $\qquad $  $\Longrightarrow$ $y$를 설명하기 위해 불필요한  Feature들을  도입한 경우 <br> $\qquad \quad $ $ \checkmark \ $ Dataset 및 Model이 불필요하게 복잡해짐  <br> $\qquad \qquad $ $ \cdots \ $ 불필요하게 높은 차원의 Data를 보유하게 됨 <br> $\qquad \quad $ $ \checkmark \  $ 차원축소 방법등을 이용하여 $z$의 component 중 불필요한  feature를 제거함 <br> <br>  $\quad$ (Case 3) $\mathcal{Z} \subset \mathcal{X}$인 경우 <br> $\qquad $  $\Longrightarrow$ $y$를 설명할 수 있는 중요한 Feature를 빠뜨린 경우  <br> $\qquad \quad $ $ \checkmark \  $ $y$에 대한 설명력이 낮은 Model을 사용하게 됨 <br> $\qquad \quad $ $ \checkmark \  $ 전문가의 자문 등 여러 정보를 수집하여 의미있는 $z_j$를 Model에 추가해야함 
<br>function  $f$ |  <br> $\bullet \ $ 선형회귀 Model: 함수 $f$를  $ \left( z_1, ..., z_{n_{_0}} \right)$의 선형 Model로 근사화 <br><br> $$
y = w_0 + w_1 z_1 + \cdots + w_{n_{_0}} z_{n_{_0}} + \varepsilon
$$ <br> $\bullet \ $  Deep Neural Network: 함수 $f$를 $ \left( z_1, ..., z_{n_{_0}} \right)$를 input으로하는 network model로 근사화 <br><br> $$ 
y = \sigma^{(L)}  W^{(L)} \cdots \sigma^{(1)}  W^{(1)} z  + \varepsilon 
$$
<br> 적절한 $ f$와 <br> 적절한 feature $z$의 <br> 도입에 대한 평가 |   <br> $\bullet \ $ $f$에 대한 model이 적절하고 feature $z = \left( z_1, ..., z_{n_{_0}} \right)$ 가 잘 선택되었다면  다음과 같이 정의되는   Loss <br> $ \ \ $ function 값은 작은 값이 될 것임 <br><br> $\qquad$ (i) MSE <br> $$ MSE \left(y, \hat y  \right) = \sum_{i=1}^{n} \left( y_i - \hat y_i  \right)^2$$ <br> $\qquad$ (ii) Cross Entropy <br> $$ CE \left(y, \hat y  \right) \ = \  E_p \left[ log \left(\frac{1}{q \left( \hat y_i \right)} \right) \right]= \sum_{i=1}^C log \left(\frac{1}{q \left( \hat y_i \right)} \right) p \left( y_i \right)$$

#### (2)   Feature 선택과 관련한 단계
#### (2.1) Dataset 및 Model과 관련한 Feature  선택
- Feature 선택은 한번에 완결되는 단계가 아닌 Data 수집부터 Model을 이용한 분석까지 각 단계와 관련이 있음

단계 | $\quad$ | Feature와의 관계
:----:|:----|:----
<br> Raw Data <br> 수집 |  | <br> $\bullet \ $ data 수집단계에서 label $y$와 동시에 수집하는 feature $z_j$들로 구성된 dataset $Z_i$를 수집  <br> $\quad$ $\checkmark$ $z_{ij}$는 dataset $Z_i$에 포함된 $j^{th}$ feature를 의미
<br> Dataset $Z$  <br> 구성 |  | <br> $\bullet \ $ 여러  dataset  $Z_i$을 통합하여 분석  dataset $Z$을 구성하는 경우 각 dataset $Z_i$에 포함된 <br> $ \ \ $ $z_{ij}$들의 공통집합 뿐 아니라 특정 dataset $Z_k$에 속하는 $z_{kj}$를 고려함
<br> Modeling 및  <br> 분석 |  | <br> $\bullet \ $ 현단계까지 고려하고 있는 $z_{ij}$에 기반한 Model을 이용하여 Model에 대한 설명도를 고려하여 <br> $ \ \ $ $z_{ij}$ 중 의미있는 $z_{ij}$와 의미없는 $z_{ij}$를 구별하는 과정을 통해 feature를 선택하게 됨

#### (2.2) Feature 선택과 관련한 분석
- 적절한 feature를 선택하기 위해서는 Model $M$와 연계하여 <br>
$ \ \ $ $\checkmark \ $ 특정 feature $z_i$가 Model에 영향을 미칠 것으로 가정한 feature set $z^{(1)}$을 구성함 <br>
$ \ \ $ $\checkmark \ $ Model $M$에 유의미한 영향을 미칠 것이라고 가정한 feature set의 각 원소인 feature $z_i$ 각각에 대해, <br> $ \quad \ \ $수집된 data를 기반으로 Model에 유의미한 영향을 미치는가를 확인하는 과정인 <span style="color:red">Feature Set Evaluation</span>이 필요  <br>
$ \ \ $ $\checkmark \ $ 유의미한 영항을 미치는 feature $z_i$들만으로 구성된 최종 feature set $z$를 구성

### 1.4 Feature Engineering

#### (1) Feature Engineering의  개념

(1) Feature Engineering은 Label  $y$와 원시 Feature $ z = \left(  z_1,  z_2, ...,  z_{n_{_0}} \right)$에 대해 <br>
(2) 다음과 같은 2종류의 문제를 해결하는데 필요한 <span style="color:red"> Feature변환과정</span>을 다루는 것을 말함 <br>
$\quad$ (i) 지도학습의 예측 Model $y = f \left(  z_1,  z_2, ...,  z_{n_{_0}} \right)$의 성능을 향상<br>
$\quad$ (ii) 비지도학습에서의 원시 Feature $ z = \left(  z_1,  z_2, ...,  z_{n_{_0}} \right)$의 숨겨진 구조($z $ Data가 Sampling된 Population에 <br> $\qquad$ 대해 알아냄)를 파악 <br>
$\qquad$ $\checkmark \ $ 여기서 Feature 변환과정이란, 원시 Feature $ z = \left(  z_1,  z_2, ...,  z_{n_{_0}} \right)$를 다양한 함수변환을  통해 

<br>
$$
\tilde z_i = g_{i} \left( z_1, z_2, ..., z_{n_{_0}} \right)
$$
<br>
$\qquad \quad$ $ \tilde z = \left( \tilde z_1, \tilde z_2, ..., \tilde z_p \right)$로 변환하는 과정을 말함<br>
$\qquad$ $\checkmark \ $ 이러한 변환과정을 Feature Learning이라고 부름 
<br>

- 모든 Machine Learning Algorithm은 Input Data $z=\left( z_1, ..., z_{n_{_0}} \right)$를 사용하여 출력 $y$을 생성함 <br>
$ \ \ $ $\checkmark \ $ Input Data $z$는 Feature들로 구성되고, 구조화된 열 형태로 표현됨<br> 
$ \ \ $ $\checkmark \ $ Machine Learning Algorithm이 제대로 작동하려면 몇 가지 특정한 특성을 가진($y$의 변화를 잘 설명하는) <br> $ \quad \ \ $ Feature들이  필요함<br><br>

- Feature들을 잘 선택하여 Model을 만들면 Data $\left( z_1, ..., z_{n_{_0}} \right)$와 $y$ Data들간의 관계 $y=f \left( z_1, ..., z_{n_{_0}} \right)$를 좀 더 정확하게 Approximation 할 수 있으므로 보다 바람직한 Model을 만들 수 있게 됨<br>
$ \ \ $ $\checkmark \ $ Feature들은  Data Source로 부터 얻어진 Feature들을 분해 또는 분할하여 새로운 Feature를 찾음

#### (2) Feature Engineering의 목표

#### $\quad$ 목표 1: Model의  예측 정확도 높이기

$\qquad$ $\bullet \ $ Feature $z$와 관측치(또는 label) $y$간의 관계$y=f \left( z_1, ..., z_{n_{_0}} \right)$를 잘 설명 하는 $\left( z_1, ..., z_{n_{_0}} \right)$의 선택을 통해, <br> $\qquad \quad $정확한 예측값  $\hat y$을 얻는 것이 목표임<br>
$\qquad \quad$ $\checkmark \ $ 즉, 다음과 같은 Optimization Problem에서 더 좋은 Solution를 구할 수 있음을 의미<br><br>

<span style="color:blue">
$$ 
\begin{align}
(Regression & Problem) \\
& \underset{W}{minimize} \ \ Loss(W) =  MSE = \frac{1}{n} \sum_{i=1}^{n} \left(y_i - \hat y_i \right)^2  \\
&subject \ to : y_i \ = \ \sigma^{(L)}W^{(L)} \cdots \sigma^{(2)}W^{(2)} \sigma^{(1)}W^{(1)}z_i^{(1)} + \varepsilon_i \\\\\\
(Classificati&on Problem) \\
& \underset{W}{minimize} \ \ Loss(W) =  CE \left(y, \hat y  \right) \ = \  E_p \left[ log \left(\frac{1}{q \left( \hat y_i \right)} \right) \right]= \sum_{i=1}^C log \left(\frac{1}{q \left( \hat y_i \right)} \right) p \left( y_i \right)  \\
& subject \ to : y_i \ = \ \sigma^{(L)}W^{(L)} \cdots \sigma^{(2)}W^{(2)} \sigma^{(1)}W^{(1)}z_i^{(1)} + \varepsilon_i
\end{align}
$$
</span>

#### $\quad$ 목표 2: Feature Data의 구조의 이해

$\qquad$ $\bullet \ $ Data $z=\left( z_1, ..., z_{n_{0}} \right)$에 내제된 속성 파악을 통해<br>
$\qquad \quad$ (i) $z$ Data를 더 잘 설명할 수 있거나 <br>
$\qquad \qquad$ $\checkmark \ $ 예를 들면 Feature Engineering을 통해 $z_i$와 $z_k$, $z_l$ 변수들 간의 관련성 $z_i = h_i \left( z_k, z_l \right)$을 알수  있음<br>
$\qquad \quad$ (ii) $z$에 포함되어 있는 더 깊은 지식을 획득할 수 있음<br>
<br>
$\qquad$ $\bullet \ $ Feature Engineering을 통해 비지도학습(Unsupervised Learning)의 성능을 올릴 수 있음

#### $\quad$ 목표 3: 분석 및 예측 과정에서 필요한 계산량 줄이기

$\qquad$ $\bullet \ $ 불필요한 Feature의 제거(차원축소와 동일한 개념임) 등에 따라 계산량을 줄일 뿐 아니라 저장공간 등을 절약 <br>

$\qquad$ $\bullet \ $ Model관점에서는 같은 성능을 가지면서 좀 더 간단한 Model을 얻을 수 있고 이에 따라 Feature와 Label간의 <br> $\qquad \quad$ 관계를 더 쉽게 이해할 수 있음

#### (3) Feature와 Dataset

- Data는 확률적으로 움직이는 변수로 data를 취할 때마다 다른 값을 가짐

- 문제의 구조를 보다 더 잘 설명하기 위한 feature를 찾아내기 위해 원시 data의 집계방법이나 분해하는 방법은 무엇일까?

- 테이블 형식 Dataset은 $z=\left( z_1, z_2, ..., z_{n_{_0}} \right)$로 표현되는 변수 또는 속성이 열( column)에 배치되고 각 $z_i$들의 조건 조합에서 얻어지는 관측치(Observation, 또는 사례 (Instance))  $y$로 구성됨<br>
$ \ \ $ $\checkmark \ $ Dataset은 각 $z_i=\left( z_{i,1}, z_{i,2}, ..., z_{i,n_{_0}} \right)$의 조건에서 관측한 $y_i$로 구성되어 있음 $\left( z_{i,1}, z_{i,2}, ..., z_{i,n_{_0}}, y_i \right)$

### 1.5 Feature Selection과 Feature Extraction

#### (1) Feature Selection

- 차원의 저주<br> 
$ \ \ $ $\checkmark \ $ 너무 많은 feature를 가질 때 발생하는 문제<br>
$ \ \ $ $\checkmark \ $ Feature 수가 많을수록 model의 표현력이 향상되지만 모든 feature가 model의 표현력과 관련이 있는  <br> $ \quad \ \ $ 것은 아닐 수 있음<br>
$ \ \ $ $\checkmark \ $ 차원이 높을수록(model을 설명하기 위해 동원된 feature의 수가 많을 수록) 무의미한 feature가 존재할  <br> $ \quad \ \ $ 가능성이 높아짐<br><br>

- 특징 선택(feature selection)은 고차원 data를 저차원 data로 변경시키는  방법 중 하나임<br>
$ \ \ $ $\checkmark \ $ Label data에 의미있는 영향을 미치는 feature 만을 선택하는 것이 핵심 개념임<br>
$ \ \ $ $\checkmark \ $ 예를 들어, 문제가 주문을 예측하는 것이라면, 주문과 관련한 모든 변수/필드를 사용하는 대신  <br> $ \quad \ \ $ 전체 feature들을 모두 사용한 예측과 비슷한 예측을 하고 작은 분산을 갖는 관련 변수/필드만 선택하는 것임<br><br>

- Feature selection을 위한 핵심 접근 방식의 예<br>
$\quad$ (i) Supervised learning 접근 방식<br>
$\quad$ (ii) Unsupervised learning 접근<br>
$\quad$ (iii) 정규화<br>
$\qquad$ $\cdot \ $ Feature에 적절한 점수를 할당한 다음 해당 점수로 순위를 매길 수 있음<br>
$\qquad$ $\cdot \ $ 점수가 가장 높은 기능을 선택할 수 있음<br><br>

- Feature 중요도 점수는 유용한 것으로 추정되는 기능과 유사하지만 새로운 기능을 추출하거나 구성하는 데 사용할 수 있는 정보를 제공할 수도 있음<br><br>

- Label $y$와 상관관계가 높은 feature는  중요한 feature 일 가능성이 높음

#### (2) 특징추출(Feature Extraction)

- 특징추출(Feature Extraction) 방법은 기존 feature들$\left( z_1, ..., z_{n_{0}} \right)$에 적절한 함수를 취하여 얻은 세로운 feature를 선택하는 방법을 말함

$$
x_i = g_i \left( z_1, ..., z_{n_{0}} \right),  \ \forall \ i
$$

- PCA, Autoencoder 등이 이 방법에 해당<br><br>

#### (2.1) PCA

- PCA는 기존 feature들의 선형결합으로 새로운 feature를 생성함 <br>

$$
x_i = \sum_{i=1}^{n_{_0}} c_iz_i
$$
$\qquad$ 여기서 $c_i$는 scalar 값 <br><br>

#### (2.2) Autoencoder

- Autoencoder는 기존 feature들의 비선형결합으로 새로운 feature를 생성함

$$
x_i = \sigma^{(k)} W^{(k)} \cdots \sigma^{(2)} W^{(2)}\sigma^{(1)} W^{(1)}z
$$

$\qquad$ 여기서 $W^{(i)}$는 $i^{th}$ layer에 사용된 선형변환 matrix이고 $\sigma^{(i)}$는  $i^{th}$ layer에 사용된 componentwise로 <br> $\qquad  \ $ 작동하는 activation function임

## 2. 표현학습(Representative Learning)

### 2.1 표현학습 개요


#### [표현학습] <br>
>(i) 표현학습은 input data를 묘사하는데 중요한 특성(feature)을 변환하거나 가공하여 data로 부터 <br> $\quad$ 특징을 추출하는 학습방법을 말함<br>
<br>
(ii) Feature learning이라고도 부름

#### (1) 표현학습이 필요한 이유

#### (예제 1)

- 다음과 같은 로마 숫자(Roman numerical representation)의 연산을 생각해보자

$$
\qquad \qquad  \qquad \qquad \qquad \qquad \frac{CCV}{V} =? \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad (1)
$$

- 로마숫자 $C$는 100을 의미하고 $V$는 5를 의미하므로 다음과 같이 아라비아숫자(Arabic numerical representation)로 <br> $ \ \ $ 바꾸어 생각하면  쉽게 문제를 풀 수 있음

$$
\qquad \qquad  \qquad \qquad \qquad \qquad \frac{205}{5} =41 \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad (2)
$$

- 문제 (1)과 문제 (2)가 같은 문제임에도 문제 (1)을 풀이하는 것이 문제 (2)를 풀이하는 것 보다 혼란스러운 <br> $ \ \ $ 이유는 문제에서 주어진  정보들을 표현하는 방식이 다르기 때문임

 $\quad$ $\Longrightarrow$ 우리에게 주어진 data의 **표현(representation)**을 바꿈으로써 더 효율적인 방법을 찾은 것임

#### (예제) MNIST Dataset의 Classification문제와 표현학습


#### (1) MNIST Classification문제

- MNIST 분류문제를 표현학습관점에서 설명해보면 다음과 같음

- 단, 각 숫자 data는 $28 \ \times \ 28$인  784개 pixel값으로 구성된 data임

- $28 \ \times \ 28$ 숫자 data를 분류하기 위해 2개의 hidden layer를 이용한 neural network을 사용하였다고 하자<br>
\- 각 hidden layer의 node수는 512과 10개로 설정한 경우임

- 지도학습 Deep Neural Network으로 순방향(forward propagation)하는 것은 일종의 표현학습(representative learning) 과정으로 볼 수 있음<br>
$\quad$ (1) 순방향 마지막 layer은 일반적으로 선형분류기(보통 softmax를 사용함)이고, <br>
$\quad$ (2) 나머지 layer들은 표현(representative)을 제공하는 방법을 학습하는 과정으로 이해할 수 있음<br>
$\quad$ (3) MNIST data를 $784차원 \ \rightarrow \ 512 차원 \ \rightarrow \ 10차원$으로 보내 분류하는 문제는 다음과 같은 과정을 따르는 것임 

> $\bullet \ $ 즉 MNIST 분류문제 풀이과정은 <br><br>
>
>$\quad$  (i) 우리의 눈으로 보이는 $28 \times 28$ pixel  image data를 <br>
$\quad$ (ii)  Flatten하여 784차원의 한 점(point)으로 나타(표현)내고 <br>
$\quad$ (iii) 784차원의 point를 행렬 $W^{(1)}$를 이용하여  512차원의 한 point로 다르게 표현하고 <br>
$\qquad$ $\checkmark$ 행렬 $W^{(1)}$와 784차원 벡터의  곱은 다음 3가지를 의미함<br>
$\qquad \quad $ $\cdots \ $ 784차원 벡터의 크기를 변화시키고 회전시키는 선형변환과<br>
$\qquad \quad $ $\cdots \ $ $W^{(1)}_{784 \times 512}$를 이용한 행렬곱은 주어진 784차원의 data를 784차원에서  512차원의 data로 <br> $\qquad \qquad$ 차원변환을 의미함<br>
$\qquad$ $\checkmark$ 또한, 이 때 사용한 $W^{(1)}$는 loss function을 최소화하는 $W$임<br>
$\quad$  (iv)  512차원 내에서 비선형변환 함수인 activation function $\sigma^{(1)}$을 이용하여 같은 차원 내의 다른 <br> $\qquad $ point로 이동하고(표현을 바꾸고)<br>
$\quad$  (v)  같은 방법으로 각 layer를 지나갈 때 (iii)과 (iv)를 반복하는 과정을 이용하는 <br>
$\quad$  (vi) data의 표현(representation)을 바꾸어 분류문제를 풀이한 것이라고 볼 수 있음

### 2.2 좋은 표현방법들(Representative Methods)

#### (1) 주어진 Data에 대해 더 좋은 표현을 찾았을 때의 장점들

$\quad$ (i) Deep neural network을 이용한 더 좋은 표현은, <span style="color:red">더 작은 Loss 값</span>을 가지게 하는 <span style="color:red">더 좋은 근사화가 실현</span>되었음을 의미<br>
$\quad$ (ii) 찾아진 좋은 표현(representation)을 이용하여, 좋은 표현을 모방한 data를 생성(generation)할 수 있음<br>
$\qquad$ $\checkmark \ $ 좋은 표현을 모방한 Data를 생성(generative)할 수 있다면 <br>
$\qquad \quad$ $\cdot \ $ <span style="color:red">지도학습(supervised learning)</span>에서 label $y$을 얻기 어려운 경우 $y$에 대응하는 많은 $z_i$를 생성함으로써 <br> $ \qquad \qquad $ 학습에 필요한 data $\left( z_{ij}, y_i \right), \ j=1, .... , n_i$를  생성할 수 있음<br>
$\qquad \quad$ $\cdot \ $ <span style="color:red">비지도학습(unsupervised learning)</span>에서 data $z_i, \ \ i=1, ..., n$을 추가 생성함으로써 dataset $Z$의 <br> $ \qquad \qquad $ 숨겨진 구조를 학습하는데 필요한  data를 확보할 수 있음 <br>
$\qquad \quad$ $\cdot \ $ <span style="color:red">강화학습(reinforcement learning)</span>에서 전이함수와 보상함수에 대한 model을 모를 때, 필요한 data인 <br> $ \qquad \qquad $episode를 생성함으로써  강화학습할 수 있음 

#### (2) 더 좋은 표현(representation)을 찾는 문제의 어려움 

- Training dataset의 분포를 모방한 설득력 있는 data를 생성하는 것은 까다로운 작업임<br><br>

- 이 문제가 까다로운데는 몇 가지 이유가 있음<br>
$\quad$ (1) 이 작업은 $z$ data 즉, unsupervised data 분포의 표현(representation)을 학습하는 것임<br>
$\quad$ (2) 즉, $k-$Means를 사용한 clustering의 경우  data  수준의 작업이지만 생성 Model(generative model)은 <br> $\qquad$ $z$ data의 숨겨진 구조(underlying  structure)를 찾는 문제임<br>
$\quad$ (3) data의 숨겨진 구조를 잘 학습했다면, 숨겨진 구조의 특성을 잘 활용하여 설득력 있는 가짜 data를 <br> $\qquad$ 생성할 수 있음<br>
$\qquad $ $\checkmark$ Data를 추가로 생성(generation)하여 지도학습, 비지도학습 및 강화학습의 성능을 향상시킬 수 있음

### [연습] Concrete Feature 예

[자료] https://www.kaggle.com/datasets/elikplim/concrete-compressive-strength-data-set


- Concrete Dataset에는 다양한 Conctrte Formulation들에 따른 concrete가 견딜 수 있는 하중을 측정한 concrete의 압축 강도가 포함되어 있음
- 이 예제 분석의 목적은 concrete formulation이 주어지면 Concrete의 압축 강도를 예측하는 것임


```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

df = pd.read_csv("c:\workspace\concrete.csv")
df.head()
```

#### Step 1
- Cement 부터  FineAggregate까지 다양한 콘크리트에 들어가는 다양한 재료를 볼 수 있음
- 다양한 재료들로부터 파생된 몇 가지 추가 합성 feature을 추가하면 Model이 이들 간의 중요한 관계를 학습하는 데 어떻게 도움이 될 수 있는지 잠시 후에 보게 될 것임<br><br>

#### Step 2
- 먼저 증강되지 않은 dataset에서 model을 훈련하여 기준 concrete압축강도를 설정함
\- 합성된 새로운 feature를 고려하지 않은 original feature로 만든 model과 성능비교를 할 것임 <br><br>

#### Step 3
- 이와 같은 기준 concrete압축강도를 구하는 것은 feature engineering process를 시작할 때 좋은 방법임
-  기준 concrete압축강도는 새 feature을 유지할 가치가 있는지 또는 폐기하고 다른 feature를 포함해야하는지 여부를 결정하는데 도움이 될 수 있음

```python
X = df.copy()
y = X.pop("CompressiveStrength") #  cement dataset에서 "CompressiveStrength"를 추출하여 y변수로 설정

# Train and score baseline model
# Model로 RandomForestRegressor()를 사용

baseline = RandomForestRegressor(criterion="absolute_error", random_state=0)
baseline_score = cross_val_score(
    baseline, X, y, cv=5, scoring="neg_mean_absolute_error"
)
baseline_score = -1 * baseline_score.mean()

print(f"MAE Baseline Score: {baseline_score:.4}")
```

- 요리에서  recipe의 재료 비율이 절대적인 양보다 recipe가 어떻게 나오는지를 더 잘 예측할 수 있다는 것을 알 수 있음
- 그러면 위의 feature들의 비율이 concrete_compressive_strength의 좋은 예측 변수가 될 것이라고 추론할 수 있음<br><br>

- 아래 셀은 세 가지 새로운 비율 feature를 Dataset에 추가함

```python
X = df.copy()
y = X.pop("CompressiveStrength")

# Create synthetic features
X["FCRatio"] = X["FineAggregate"] / X["CoarseAggregate"]
X["AggCmtRatio"] = (X["CoarseAggregate"] + X["FineAggregate"]) / X["Cement"]
X["WtrCmtRatio"] = X["Water"] / X["Cement"]

# Train and score model on dataset with additional ratio features
model = RandomForestRegressor(criterion="absolute_error", random_state=0)
score = cross_val_score(
    model, X, y, cv=5, scoring="neg_mean_absolute_error"
)
score = -1 * score.mean()

print(f"MAE Score with Ratio Features: {score:.4}")
```

$\Longrightarrow$ 확실히 성능이 향상되었음<br>
$\quad$ 이것은 이러한 새로운 ratio features 이전에 감지하지 못했던 중요한 정보를 Model에 포함되었다는  증거임

### [연습] Automobile Dataset

[자료]  https://www.kaggle.com/code/alephvnull/automobile-price-modeling-with-custom-estimator/notebook

(1) 새로운 dataset에는 충분한 설명 없는  수백 또는 수천 개의 feature가 포함되어 있을 수 있음<br><br>

(2) 첫 번째 단계는 feature $z = \left( z_1, ..., z_{n_{_0}} \right)$와 label(또는 target) $y$ 간의 연관성을 측정하는 기능인  feature utility <br> $\quad$ metric으로 순서를 매기는 것임<br>
- 그런 다음 feature들 중에  가장 유용한 feature들($y$와 연관성이 큰 feature들)로 구성된  작은 dataset을 구성하는 것임<br><br>

(3) 우리가 사용할 metric을 "상호 정보(mutual information, MI)"라고 부름<br>
$\qquad \quad$ [참고] 정보이론(information theory)

$$
\begin{align}
MI(X, Y) =& D_{KL} \left( p_{_X, _Y} (x, y) \| p_{_X} (x) p_{_Y} (y) \right) \\
        =& E_{p(_X, _Y)} \left( log \frac{p(X, Y)}{p(X)p(Y)} \right)  \\
        =& \sum_{y \in Y} \sum_{x \in X} log \frac{p(x, y)}{p(x)p(y)} p(x, y)
\end{align}
$$

- 결합확률분포와 각 주변확률분포의 곱의 차이를 보는 measure임<br>
\- 확률변수 $X$와 $Y$의 독립성을 보는 measure로 볼 수 있음<br><br>

- 상호 정보(mutual information)는 2개의 변수 간의 관계를 측정한다는 점에서 상관관계(correlation)와 매우 유사함
- 상호정보의 장점은 모든 종류의 관계를 감지할 수 있는 반면 상관 관계는 선형 관계만 감지한다는 점임
- 상호 정보는 훌륭한 범용 메트릭이며 아직 어떤 Model을 사용하고 싶은지 모를 때 기능 개발을 시작할 때 특히 유용함
