- Linear Regression이 해결할 수 있는 문제에 대한 이해
- 상관관계의 적용범위
- Gradient Descent와 관련한 수학지식

## 1.  Regression과 Machine Learning의 관계

- Regression은 지도학습 방법 중 한 가지 방법임

#### 다양한 궁금증

(i) 어린이들의 키 $y$는 왜 차이가 있을까?<br>
$\quad$ $\checkmark \ $ 아빠의 키 $z_1$은 어린이의 키$y$에 영향을 미칠까?<br>
$\quad$ $\checkmark \ $ 아빠의 키 $z_1$ 이외에 1일 영양섭취량 $z_2$도  어린이의 키$y$에 영향을 미칠까?<br>
$\quad$ $\checkmark \ $ $\qquad$ $\vdots$  $\qquad \qquad$ $\vdots$  <br>


(ii) 이렇게 우리가 생각하는 변수들 $ \left( z_1, z_2, ..., z_{n_{_0}} \right)$간에 어떤 관계가 있을까?<br>
$\quad$ $\checkmark \ $ $y$에 영향을 미칠 것으로 예상되는 변수들은 다음과 같다

$$
y = f \left( z_1, z_2, ..., z_{n_{_0}} \right)
$$

$\quad$ $\checkmark \ $ $y$에 영향을 미칠 것으로 예상되는 $z_1, z_2, ..., z_{n_{_0}}$간에는 어떤 관계가 있을까?

$$
z_i = g_i \left( z_1, ..., z_{i-1}, z_{i+1}, ...,  z_{n_{_0}} \right)
$$


(iii) $z_i$와 $z_j$간에 함수 관계가 존재할 때 Linear Regression은 어떤 영향을 받을까?


## 2.  Linear Regression(선형회귀분석) Model과 가정


-  Unknown function $y = f \left( x_1, x_2, ..., x_{K} \right)$에 대해 다음과 같은 가정을 함<br>

가정 | $\qquad \qquad \qquad$ 설명
:----:|:----
<br> Feature 변수 | <br> $\bullet \ $ $y$에 영향을 미칠 것이라고 생각하는 <span style="color:red"> $n_0$개의 feature 변수를 가정함</span> $$
\left( z_1, z_2, ..., z_{n_{_0}} \right)
$$
<br> Model  | <br> $\bullet \ $  $y$에 영향을 미칠 것이라고 생각하는 feature들과 <span style="color:red">선형관계라고 가정</span>함 $$
y = w_0 + w_1 z_{1} + w_2 z_{2} + ...  + w_{n_{_0}} z_{n_{_0}} + \varepsilon_i, i=1, 2,...,n
$$
<br> Model의 오차항 $\varepsilon$ | <br> $\bullet \ $ Model의 <span style="color:red">오차항 $\varepsilon$에 대해 다음과 같이 가정함</span> $$\begin{align} 
& (1) \ E(\varepsilon_i) = 0, \ \ i=1, ..., n  \\
& (2) \ Var(\varepsilon_i) = \sigma^2, \ \ i=1, ..., n
\end{align}$$ 


## 3.  Linear Model Parameter $w= \left(w_0, w_1, \cdots, w_{n_{_0}}  \right)$ 의 추정

- Dataset $(z_i, y_i), i=1,...,n$를 이용하여 $ \widehat{w} $을 추정(estimation)하는 방법으로 2가지 방법이 있음<br>

$\quad$ (i) Analytical Method<br>
$\quad$ (ii)  Gradient Descent Method<br>

###  3.1  Analytical Method를 이용한  Regression Parameter $W$ 추정

#### (1) 단순회귀분석의 parameter추정

1. Feature 변수가 1개인 간단한 Linear Regression Model을 생각해보자<br>

$$
    y_i = w_0 +w_1 z_i + \varepsilon_i, i=1,..., n
$$ 

2. Model parameters의 추정의 대상은 $w_0$, $w_1$과 $\sigma^2=Var(\varepsilon_i)$임<br>
$\quad$ $\checkmark \ $  **mean squared errors (MSE)** 또는 **Error Sum Squares(SSE)** 을 최소화하는 parameter를 구하는 것임<br> 
$\quad$ $\checkmark \ $ 인공지능에서는 MSE와 같은 목적함수(objective function)을 Loss function이라고 부름<br>
$\quad$ $\checkmark \ $ 이렇게 구해진 parameter의 추정치를  **Ordinary Least Squares (OLS)** 라고 부름<br>
<br>

<span style="color:blue">
$$
\begin{align}
minimizes \ \ &Loss(w_0, w_1) = \sum_i^{n} \varepsilon_i^2 \\
subject \ \ to \ \   &y_i = w_0 + w_1 z_i + \varepsilon_i, i=1,...,n
\end{align}
$$
</span>
<br>

3. $Loss$를 최소화하는 $w_0$와 $w_1$를 찾기 위해 다음을 만족하는 $\hat w_0$와 $\hat w_1$를 구하면 됨<br>

$$
\begin{align}
\frac{\partial Loss }{\partial w_0} =& 0  \\
\frac{\partial Loss }{\partial w_1} =& 0
\end{align}
$$

$ \ \ $ 3.1 $w_0$로 $Loss$를 편미분 해보면:

$$
\frac{\partial Loss}{\partial w_0} = \frac{\partial}{\partial w_0} \sum_i(y_i - w_0 - w_1 z_i)^2 = 0\\
\sum_i y_i = \hat w_1 \sum_i z_i + n \hat w_0\\
n \bar{y} = n \hat w_1 \bar{z} + n \hat w_0\\
\hat w_0 = \bar{y} - \hat w_1 \bar{z}
$$<br>

$ \ \ $ 3.2 $w_1$로 $Loss$를 편미분 해보면:

$$
\frac{\partial Loss}{\partial w_1} = \frac{\partial}{\partial w_1} \sum_i(y_i - w_0 - w_1z_i)^2 = 0
$$
<br>

$\quad$ 위에서 구한 $\hat w_0$를 대입하면:
<br>
$$
\sum_i z_i(y_i - \hat w_1 z_i - \bar{y} + \hat w_1 \bar{z}) = 0\\
\sum_i z_i y_i - \bar{y}\sum_i z_i = \hat w_1 \sum_i(z_i - \bar{z})
$$
<br>

$\quad$ 양변을 $n$으로 나누면:

<br>
$$
\frac{1}{n}\sum_i z_i y_i  - \bar{y}\bar{z} = \frac{1}{n} \hat w_1 \sum_i(z_i - \bar{z})\\
\hat w_1 = \frac{\frac{1}{n}\sum_i z_i y_i  - \bar{y}\bar{z}}{\frac{1}{n}\sum_i(z_i - \bar{z})} = \frac{S_{zy}}{S_{zz} }
$$

#### (2) 벡터 표현을 이용한 Parameter 추정

1. Linear Regression Model<br>
$\bullet \ $ $n_{_0}$개의 independent variable $z = \left( z_1, z_2, ..., z_{n_{_0}} \right)$이 있는 model<br>

$$
y = w_0 + w_1 z_{1} + w_2 z_{2} + ...  + w_{n_{_0}} z_{n_{_0}} + \varepsilon
$$

$\qquad \quad$ 여기서  $E(\varepsilon)=0$, 이고 $Var(\varepsilon)=\sigma^2(constant \ variance)$

2. $n$개 data $ \left(z_i, y_i \right), \ i=1, ..., n$을  Linear Regression model에 대입해 보면 다음과 같음 

$$
\begin{align}
\qquad y_1 &= w_0 + w_1 z_{1,1} + w_2 z_{1,2} + ...  + w_{n_{_0}} z_{1,n_{_0}} + \varepsilon_1 \  \rightarrow \ 1^{st} \ data \left( y_1, z_{1,1}, z_{1,2}, ..., z_{1,n_{_0}}  \right)  \\
       y_2 &= w_0 + w_1 z_{21,} + w_2 z_{2,2} + ...  + w_{n_{_0}} z_{2,n_{_0}} + \varepsilon_2 \  \rightarrow \ 2^{nd} \ data \left( y_2, z_{2,1}, z_{2,2}, ..., z_{2,n_{_0}}  \right) \\
       \ \ \vdots \  & \qquad \qquad \vdots \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \vdots\\
       y_n &= w_0 + w_1 z_{n,1} + w_2 z_{n,2} + ...  + w_{n_{_0}} z_{n, n_{_0}} + \varepsilon_n \  \rightarrow \ n^{th} \ data \left( y_n, z_{n,1}, z_{n,2}, ..., z_{n,n_{_0}}  \right) 
\end{align}   
$$


3. 행렬, 벡터 표현

$\quad$ $\checkmark \ $ 위의 $n$개의 regression model을 벡터로 표현하면 다음과 같음<br><br>

<span style="color:blue"> 
$$ 
\underset{n \times 1}{y} = \underset{ n \times \left( n_{_0}+1 \right) }{Z} \underset{ \left( n_{_0}+1 \right) \times 1 }{w} + \underset{n \times 1}{\varepsilon}
$$
</span>

$\qquad$ $\cdots$ Matrix $Z$와 vector $w$의 size에 $ \left( n_0 + 1 \right)$인 이유는 feature변수 $z_i$에 대응하는 $w_i$ parameter가  $n_0$개 있고 <br> $\qquad \quad$ 절편인 $w_0$가 있기 때문음
<br><br>
$\qquad \qquad$ 여기서, 
$ 
y = 
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n \\
\end{bmatrix}
$, $\quad $
$
Z =
\begin{bmatrix}
1 \ z_{1,1} \ z_{,12} & \cdots & z_{1,n_{_0}} \\
1 \ z_{2,1} \ z_{2,2} & \cdots & z_{2,n_{_0}} \\
\vdots \ & \ddots & \vdots \\
1 \ z_{n,1} \ z_{n,2} & \cdots & z_{n,n_{_0}} \\
\end{bmatrix}
$, $\quad $
$
w = 
\begin{bmatrix}
w_0 \\
w_1 \\
\vdots \\
w_{n_{_0}} \\
\end{bmatrix}
$, $\quad $
$
\varepsilon = 
\begin{bmatrix}
\varepsilon_1 \\
\varepsilon_2 \\
\vdots \\
\varepsilon_n \\
\end{bmatrix}
$

$\qquad \qquad \qquad \qquad \ \   n \times 1  \qquad \qquad \qquad  n \times (n_{_0}+1) \qquad \qquad \quad (n_{_0}+1) \times 1 \qquad  n \times 1$

4.  선형회귀모델의 Parameter 추정문제(Linear Regression Parameter Estimation Problem)<br>
- Quadratic Programming임<br>
$ \ \ $ $\checkmark \ $ Objective function: Quadratic form(목적함수에 우리가 추정하고자하는 parameter의 2차항을 포함하고 있음)<br>
$ \ \ $ $\checkmark \ $ Constraints: Linear form(제약식은 1차항만으로 구성)<br><br>
<br>
<span style="color:red">
$$
\begin{align}
 & Objective \ \ Function \ \ \  \ \  \underset{w}{minimize} \ \ Loss(w) =  \varepsilon'\varepsilon = (y-Zw)'(y-Zw)  \\
 & subject \ to \qquad \qquad \quad  y = Zw + \varepsilon  
\end{align} 
$$ 
 </span>
<br>

5. $\frac{\partial Loss(w)}{\partial w}=0$를 만족하는 $\widehat w$를 구하면 됨<br>

$$\qquad \qquad 
\begin{align}
 Loss(w) =&  \varepsilon'\varepsilon \\
      =& (y-Zw)'(y-Zw) \\
      =& y'y-2Z'yw+w'Z'Zw \\\\
\frac{\partial Loss(w)}{\partial w} =& -2Z'y+2Z'Zw \qquad \qquad \qquad \qquad (1) \\\\
만일 \ \left( Z'Z \right)^{-1}가& \ 존재한다면  \\\\
\widehat{w} =& (Z'Z)^{-1}Z'y
\end{align}
$$

$\qquad$ $\checkmark \ $ Unknown parameter $w = \left( w_0, w_1, w_2, ..., w_{n_{_0}} \right)$를 Data $(Z, y)$를 이용하여  $\hat w =\left( Z'Z \right)^{-1}Z'y$로 추정한 것임

6.  $\widehat{w} = (Z'Z)^{-1}Z'y $의 기하학적 해석<br>

- 우리가 구한 Loss를 최소화하는 추정치 $\widehat w$는 Least Squares Estimator라고 부름 <br>
$ \ \ $ $\because \ $ $\underset{W}{minimize} \ \  \varepsilon' \varepsilon$로 구한 추정값이라서 Least Squares Estimator라고 부름


- 우리가 구한 $\hat w$는 5의 (1)번식이 0인  $w$를 구한 것임. 즉 다음을 만족하는 식임<br>

$$
\begin{align}
& Z'y-Z'Z \widehat w = 0 \\
& Z'(y-Z \widehat w) = 0 \\\\
& \Longrightarrow Z'_{\left( n_{_0} \times 1 \right) \times n  } \bot \left(y_{n \times 1}-Z_{n \times \left( n_{_0} \times 1 \right) } \widehat w_{\left( n_{_0} \times 1 \right)} \right) \quad (\because 내적이 \ 0 \rightarrow 90^0)
\end{align} 
$$

$ \ \ $ $\checkmark \ $ $Z \hat w$의 의미를 생각해 보자

$$
\begin{align}
Z \hat w =& 
\begin{bmatrix}
z_{1, 0}  &  z_{1,1}          & z_{1,2}     & ...        &z_{1,n_{_0}}      \\
z_{2, 0}  &  z_{2,1}          & z_{2,2}     & ...        &z_{2,n_{_0}}      \\
\vdots    &  \vdots           & \vdots      & \ddots     & \vdots           \\
z_{n, 0}  &  z_{n,1}          & z_{n,2}     & ...        &z_{n,n_{_0}}
\end{bmatrix}
\begin{bmatrix}
w_{0}     \\
w_{1}     \\
\vdots     \\
w_{n_{_0}}     
\end{bmatrix}  \\\\
=& \left( z_0, z_1, ..., z_{n_{_0}} \right) \hat w \qquad \qquad \qquad \qquad \qquad (z_k는 Z행렬의 (k+1)^{th}column \ vector ) \\\\
=& \sum_{k=0}^{n_{_0}} w_k z_k 
\end{align}
$$

$\qquad$ $\cdots \ $ 즉,  $Z \hat w$는 $Z$의 column vector들을 scalar $\hat w_k$로 하여   Linear Combination으로 생성되는 vector임

$ \ \ $ $\checkmark \ $ 즉, 우리가 찾은 $\hat w$는 <br>
$\qquad$ (i) $Z$의 column vector들에 의해 생성되는  공간 $H$와 <br>
$\qquad$ (ii) $(𝑦−Z𝑤)$ vector가 수직이 되는  $\hat w$를 찾는 문제였음

###  3.2 오차(Error, $\varepsilon$)과 잔차(Residual, $r$)

- 기계학습에서 사용하는 오차와 잔차의 개념은 완전히 다른 개념임

용어 |  $\quad$ 설명 | $\quad$ 관련수식
:----:|:----|:----
<br> Error <br> $(\varepsilon)$ | <br> $\bullet \ $ 번역: 오차 <br> $\bullet \ $  오차 = 관측치 $-$ Model 기대치  <br> $ \ \ $ $\checkmark \ $  Model에서 1차 근사화 때문에 도입된 개념 |  $$
\begin{align} 
\varepsilon =& y - E(y)\\
y =& w_0 + w_1 z_{1} + w_2 z_{2} + ...  + w_{n_{_0}} z_{n_{_0}} + \varepsilon  \\
E(Y) =& w_0 + w_1 z_{1} + w_2 z_{2} + ...  + w_{n_{_0}} z_{n_{_0}}
\end{align}$$
<br> Residual <br> $(r)$ | <br> $\bullet \ $ 번역: 잔차 <br>  $\bullet \ $ 잔차 = 관측치 $-$ 추정치 <br> $ \ \ $  $\checkmark \ $ 오차를 확인할 수 없기 때문에 대안으로 도입된 개념 | $$
\begin{align}
r_i =& y_i - \hat y_i \\
    =& y_i - z_i \hat w \\
    =& y_i - z_i (Z'Z)^{-1}Z'y
\end{align}
$$ 

###  3.3 탐색적 방법를 이용한  Regression Parameter $W$ 추정

#### (1) 문제의 정의
<span style="color:blue">
$$
\begin{align}
minimizes \ \ &Loss(w) = \varepsilon'\varepsilon = \sum_i^{n} \varepsilon_i^2 \\
subject \ \ to \ \   &y_i = w_0 + w_1z_{i,1} + \cdots + w_{n_{_0}}z_{i, n_{_0}} + \varepsilon_i, \quad i=1,...,n
\end{align}
$$
</span>

$$
\begin{align}
\frac{\partial Loss(w)}{\partial w_k} =& \frac{\partial}{\partial w_k} \sum_{i=0}^{n} \left( y_i - \left( w_0 + w_1z_{i1} + \cdots + w_{n_{_0}}z_{i, n_{_0}} \right) \right)^2 \\
  =& -2  \sum_{i=0}^{n} \left( y_i - \left( w_0 + w_1 z_{i1} + \cdots + w_{n_{_0}}z_{i, n_{_0}} \right) \right) \times z_{ik}
\end{align}
$$

#### (2) 반복적인 탐색방법(Gradient Descent Method) 

- Gradient Descent 방법을 이용하여 $\frac{\partial Loss(w)}{\partial w_k}=0$를 구함

$$
w_k = w_k - \rho \left( \frac{\partial Loss(w)}{\partial w_k} \right), \quad k=0, 1, ..., n_0
$$

$\quad$ 여기서 $\rho$는 학습율로 매우 작은 양수값, $\rho=0.01$ 등
<br><br>

$$
w^* = \left( w_0^*,  w_1^*, ...,  w_{n_0}^* \right)
$$

## 4. Linear Model을 이용한 추정(Prediction) 

### 4.1 $\widehat w$을 이용한 예측  $\widehat y$

-  Feature 값이 $z_k$일 때 $y$의 예측값 $y_k$는  다음과 같이 예측할 수 있음

$$
\begin{align}
\widehat y_k =& z_k \widehat w  \\
             =& z_k \left( Z'Z \right)^{-1}Z'y
\end{align}
$$

### 4.2 예측값의 정확도 평가 

- Dataset $\left( Z, y \right) $일 때 dataset을 다음과 같이 2개의 dataset으로 구분
- Training Dataset $\left( Z_{Tr}, y_{Tr} \right)$
- Test Dataset $\left( Z_{Te}, y_{Te} \right)$

- Training Dataset $\left( Z_{Tr}, y_{Tr} \right)$를 이용하여 $\widehat w$를 추정하고 

$$
\widehat w = (Z'Z)^{-1}Z'y
$$

- Test Dataset $\left( Z_{Te}, y_{Te} \right)$를 이용 추정한 Regression Model의 성능을 여러가지 방법으로 평가할 수 있음<br><br>

#### (1) $\hat w$의 Variance

$$
\begin{align}
Var(\widehat w) =& Var \left( (Z'Z)^{-1}Z'y \right) \\
                =& \left( (Z'Z)^{-1}Z' \right) Var(y) \left( (Z'Z)^{-1}Z' \right)' \\
                =& \left( (Z'Z)^{-1}Z' \right) \sigma^2 I \left( (Z'Z)^{-1}Z' \right)' \\
                =& (Z'Z)^{-1}\sigma^2 
\end{align}
$$

#### (2) 관측치 Label과 예측치의 차이에 의한 Loss

- Test dataset $\left( z_i, y_i \right)$에 대해서

$$
\hat y = Z \hat w = Z  \left(Z'Z \right)^{-1}Zy
$$

$$
\begin{align}
Loss \left( y, \hat y \right) =& \sum_{i=1}^{k} \left( y_i - \hat y_i \right)^2 \\
                              =& \sum_{i=1}^{k} \left( y_i - z_i \left(z_i'z_i \right)^{-1}z_iy \right)^2
\end{align}
$$

## 5. Regression Model의 변동분해

### 5.1 Regression Model과 변동의 의미

- Sampling을 통해 얻은 $y_1, y_2, ..., y_{n})$들은 $y_i$의 표본평균$ \left( Sample \ Mean, \ \bar x \right)$으로 부터의 차이에 대한 제곱합은 다음과 같음

$$
SST = \sum_{i=1}^{n} \left( y_i - \bar y  \right)^2
$$

$\qquad$ $\checkmark \ $ 이러한 변동을 설명할 수 있을까?<br>



- 우리가 가정한 Linear Regression Model이 이러한 data들의 변동을 잘 설명하는지 확인하고 싶음<br> 
$ \ \ $ $\checkmark \ $ 우리가 도입한 feature변수들 $z_1, z_2, ..., z_{n_0}$와 가정한 Linear Regression Model이 data들의 변동을 잘 설명하는지 <br> $\quad \ \ $ 어떻게 확인할 것인가?<br> 


- 관측 data $\left( y_1, y_2, ..., y_n \right)$에서 각 data $y_i$는 표본평균 $\bar y$으로 부터의 총변동(Total Sum of Squares)이 Linear Regression Model에 의해 어떻게 설명되는지를 알아봄<br>
$ \ \ $ $\checkmark \ $ 가정한 model이 좋은 model이라면 Error $ \left( \varepsilon_i \right)$가 작을 것임<br>
$ \ \ $ $\checkmark \ $ $ \varepsilon_i >0$일 수도 있고, $ \varepsilon_i \le 0$일 수도 있으므로 단순한 Error의 합은 사용할 수 없음<br>
$ \ \ $ $\checkmark \ $ 전체 data들에 의한 총변동을 다음과 같이 2가지 방법으로 생각할 수 있음<br>

$$
Total \ Variation =
\begin{cases} 
\sum_{i=1}^{n} \varepsilon_i^2,     &        \ Total \ Sum \ of \ Squares  \\\\
\sum_{i=1}^{n} | \varepsilon_i |,   &        \ Total \ Sum \ of \ Absolutes
\end{cases}
$$

- Data의 변동은 평균으로부터의 차이에 대한 제곱합을 의미함<br>

$$
\begin{align}
SST =& \sum_{i=1}^{n} \varepsilon_i^2  = \sum_{i=1}^{n} \left( y_i - E(Y) \right)^2  \\\\
E(Y) =& \sum y_i P \left( Y=y_i \right)  
\end{align}
$$

- 다음 그림은 data를 Linear Regression으로 modeling한 것임 

- $y_i$들에 의한 총변동은 다음과 같이 분해할 수 있음<br>
$ \ \ $ $\checkmark \ $ $SST: \ Total \ Sum \ of \ Squares$  <br>
$ \ \ $ $\checkmark \ $ $SSR: \ Regression \ Sum \ of \ Squares$  <br>
$ \ \ $ $\checkmark \ $ $SSE: \ Error \ Sum \ of \ Squares$  <br>

$$ \begin{align}
\underbrace{ \sum_{i=1}^n(y_i - E(Y))^2}_{SST} &= \sum_{i=1}^n \left(  y_i - \hat y_i + \hat y_i - E(Y) \right)^2 \\
                             &= \underbrace{ \sum_{i=1}^n  \left( y_i - \hat y_i \right)^2}_{SSE} 
                              + \underbrace{ \sum_{i=1}^n \left( \hat y_i - E(Y) \right)^2}_{SSR} \\\\
\underbrace{SST}_{총변동} \quad &= \underbrace{SSE}_{설명 \ 불가능 \ 변동} \qquad + \qquad \underbrace{SSR}_{Regression \ Model을 \ 통해 \ 설명 \ 가능한 \ 변동} 
\end{align}$$
<br>

### 5.2 Regression Model의 적합성 척도$\left( R^2 \right)$

$\quad$ $\bullet \ $ Regression model에 포함된 독립변수 $\left( z_1, ..., z_{n_{_0}} \right)$은 우리가 가진 사전 지식에 근거하여 선정한 것임<br>
$\qquad$ $\checkmark \ $ 즉 $z_i$ 중에는 관측치 $y_i$에 영향을 미치는 변수가 있고 영향을 미치지 않는데 우리의 사전지식의 한계로 <br> $\qquad \quad$ model에 포함되는 변수로  선정되었을 수도 있음<br>

$\quad$ $\bullet \ $ Linear Regression Model을 unknown 함수 $f$를 선형모델인 $y = w_0 + w_1 z_1 + ... + w_{n_{_0}}z_{n_{_0}} + \varepsilon$으로 가정한 것임<br>
$\qquad$ $\checkmark \ $ 이 가정은 1차항까지만 근사화한 것임<br>

$\quad$ $\bullet \ $ 우리의 가정이 적합한 것이었는지 확인이 필요함<br>
$\qquad$ $\checkmark \ $ $R^2$ measure는 이러한 적합성을 판단하기 위한 measure임


$$ 
\begin{align}
& R^2 = \frac{SSR}{SST} \\\\
& 0 \le R^2 \le 1
\end{align}
$$ 

$\qquad$ $\checkmark \ $ $R^2$ 값이 1에 가깝다는 의미는 data의 총변동($SST$) 중 Linear Regression Model에 의해 설명되는 변동 ($SSR$) <br> $\qquad \quad$ 값이 크다는 의미로  Linear Regression Model이 적절하다는 의미임

$R^2$  |  | $\qquad \qquad \qquad$ 의미
:----: | :----| :----
<br> $R^2 = 0$ |  | <br> $\bullet \ $ Regression Model에 도입한 feature 변수 $z_i$들이 <br> $ \ \ $ $y_i$ data의 변동을  설명하지 못함
<br>$0 < R^2 < 1$ | |  <br> $\bullet \ $ Regression Model에 도입한 feature 변수 $z_i$들이 <br> $ \ \ $  $y_i$ data 변동의 일부를 설명$(SSR)$하고 <br> $ \ \ $ 일부는 설명하지 못함$(SSE)$  <br> $\bullet \ $ 하지만 Linear Regression Model의 도입으로 $y$의 <br> $ \ \ $  변동을 부분적으로 설명하게 됨
<br>$R^2 = 1$ | | <br> $\bullet \ $ Regression Model에 도입한 feature 변수 $z_i$들이 <br> $ \ \ $   $y_i$ data의 변동을 100% 설명함

### 6.1 해석학적 방법을 이용한 Parameter추정

```python
# direct solution to linear least squares
import numpy as np
from numpy import array
from numpy.linalg import inv
import matplotlib.pyplot as plt
%matplotlib inline

data = np.array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.50, 0.49]
])

# split into inputs and outputs
Z, y = data[:, 0], data[:, 1]
```

$$
\begin{align}
 & y = w_0 + w_1z_1+ \cdots + w_{p-1}z_{p-1} + \varepsilon     \qquad \qquad \qquad \ \ \  \ (1) \\ \\
 &\underset{w}{minimize} \ \ Loss(w) =  \varepsilon'\varepsilon    \qquad \qquad \qquad \qquad \qquad  \quad (2)  \\
 &subject \ to  \ \  y = Zw + \varepsilon   \\ \\
 &\hat w = (Z'Z)^{-1}Z'y                       \qquad \qquad    \qquad \qquad \qquad  \qquad \quad \ \  (3)
\end{align}
$$

```python
Z = np.c_[np.ones((5, 1)), Z]
print(Z.shape)
w = inv(Z.T.dot(Z)).dot(Z.T).dot(y)
w = w.T
w
```

- $y$ 예측치 계산

```python
yhat = Z.dot(w.T)
yhat
```

- $y$에 대한 $Regression \ \ Line$과 $Observed \ value$ 그리기

```python
plt.plot(Z[:,1], yhat, '-r')
plt.plot(Z[:,1], y, 'ob')
plt.show()
```

### 6.2  Gradient Descent 방법을 이용한 Parameter추정

[참고] 00_03_01 Gradient Descent (기본) .ipynb <br>
$\qquad$ 00_03_02 Gradient Descent (Dataset).ipynb <br>
$\qquad$ 00_03_03 Gradient Descent 변형들.ipynb

```python
# Making the imports
import numpy as np
import pandas as pd
import numpy.random as rnd

import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (8.0, 6.0)

# Preprocessing Input data
#X = 3*rnd.rand(100, 1) # X=100개의 난수를 생성하여 3를 곱한 수
X = 3*rnd.rand(100, 1) # X=100개의 난수를 생성하여 3를 곱한 수
y = 2 + 4*X + rnd.randn(100, 1)  # 생성된 난수 X와 정규분포를 이용해서  y data를 생성
plt.scatter(X, y)
plt.show()
```

```python
# Building the model
w = 0
w0 = 0

rho = 0.01  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    y_pred = w*X + w0  # The current predicted value of Y
    D_w = (-2/n) * sum(X * (y - y_pred))  # Derivative wrt m
    D_wo = (-2/n) * sum(y - y_pred)  # Derivative wrt c
    w = w - rho * D_w  # Update m
    w0 = w0 - rho * D_wo  # Update c
    
print (w, w0)
```

```python
# Making predictions
y_pred = w*X + w0

plt.scatter(X, y) 
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')  # regression line
plt.show()
```
