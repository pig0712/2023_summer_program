```python
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('./images/parrot.png')

im = im.resize( (50, 30) )
im = im.resize( (453, 340), resample=Image.NEAREST )
print(im.width, im.height, im.mode, im.format, type(im))

plt.imshow(im)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('./images/parrot.png')
im2 = im.resize( (50, 30) )
im3 = im2.resize( (453, 340), resample=Image.NEAREST )

plt.figure(figsize=(10,5))
plt.subplot(1,3,1); plt.imshow(im)
plt.subplot(1,3,2); plt.imshow(im2)
plt.subplot(1,3,3); plt.imshow(im3)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('./images/parrot.png')

r, g, b = im.split()

plt.figure(figsize=(6,6))
plt.subplot(2,2,1); plt.imshow(im)
plt.subplot(2,2,2); plt.imshow(r, cmap='gray')
plt.subplot(2,2,3); plt.imshow(g, cmap='gray')
plt.subplot(2,2,4); plt.imshow(b, cmap='gray')
```
