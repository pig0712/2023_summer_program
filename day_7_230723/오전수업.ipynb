```python
import matplotlib.pyplot as plt
import numpy as np

a = np.eye(128)
a *= 255

a = np.zeros((256,256))

# 최적화 하기
n = 2048
c = 0
while True:
    a[:,5+c:10+c] = n
    c += 10
    n /= 2

    if n == 1:
        break

plt.figure(figsize=(4,4))
plt.imshow(a, cmap="gray") # 색 지정
plt.figure(figsize = (100,100)) # 그림 사이즈 지정
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.eye(128)
a *= 255

a = np.zeros((256,256,3))
a[:,20:40,0] = 255
a[20:40,:,1] = 255
a[30:40,:,2] = 255


plt.figure(figsize=(4,4))
plt.imshow(a) # 색 지정
# plt.figure(figsize = (100,100)) # 그림 사이즈 지정
```

```python
from skimage import data
from skimage import color
import matplotlib.pyplot as plt

a = data.astronaut()
a = color.rgb2gray(a)

a[a>235] = 255
a[a<=235] += 20

print(type(a), a.shape)
plt.figure(figsize=(5, 5))
plt.imshow(a, cmap="gray")
```

```python
from skimage import data
from skimage import color
import matplotlib.pyplot as plt

a = data.astronaut()
a = color.rgb2gray(a)
b = a.copy()

a[a>235] = 255
a[a<=235] += 100

print(type(a), a.shape)

plt.figure(figsize=(5, 5))

plt.subplot(1,2,2)
plt.imshow(a, cmap="gray")

plt.subplot(1,2,2)
plt.imshow(b, cmap="gray")
```
