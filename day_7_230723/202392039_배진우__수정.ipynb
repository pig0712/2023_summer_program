# 문제2

```python
# 다음 리스트를 출력된 결과처럼 될 수 있도록 리스트 함수를 써서 중간과정을 채우시오
# 각 문제의 순서에 맞춰 답안 작성해야 같은 결과가 나옴

p = [1, 2, 4, 5]
q = [11, 12, 13]

print(p)   # 출력결과 : [1, 2, 4, 5]

#문제1 : p와 q를 합치기

p += q
print(p)   # 출력결과 : [1, 2, 4, 5, 11, 12, 13]

#문제2 : 값 3 추가

p.append(3)
p.sort()
print(p)   # 출력결과 : [1, 2, 3, 4, 5, 11, 12, 13]

#문제3 : 값 6 추가

p.append(6)
p.sort()
print(p)   # 출력결과 : [1, 2, 3, 4, 5, 6, 11, 12, 13]

#문제4 : 값 4 제거

p.remove(4)
print(p)   # 출력결과 : [1, 2, 3, 5, 6, 11, 12, 13]

#문제5 : 전체 순서를 거꾸로

p.reverse()
print(p)   # 출력결과 : [13, 12, 11, 6, 5, 3, 2, 1]

#문제6 : q의 값을 3회 반복

q *= 3
print(q)   # 출력결과 : [11, 12, 13, 11, 12, 13, 11, 12, 13]
```

# 문제6

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('taxis')  # 택시 운행 정보

# total 요금이 가장 큰 금액은
print(df["fare"].max(),"\n")

# pickup_borough 별 탑승자 수
print(df["pickup_borough"].value_counts())

# pickup_borough 별 total 요금의 평균 ###### 수정했습니다. #######
print(df.groupby(df['pickup_borough'])['total'].mean())

# pickup 컴럼의 자료형 보기  ##### 수정했습니다. #####
print(df.columns.dtype)

# 4일 아침 8시에 탑승 목록
df["pickup"] = df["pickup"].astype("str")
df[df["pickup"].str.contains("04 08")]

# pickup_zone 이 City로 끝나는 목록   #### 수정했습니다. ####
df[df["pickup_zone"].str.endswith("City") >= 1]


##### 빠진 문제 입니다 ######

# distance가 30이상인 목록
df[df["distance"]>= 30]

# 거리(distance)가 30 초과인 목록 모두와 요금(fare)가 140 이상 목록 모두 출력
df[(df["distance"] > 30) & (df["fare"] >= 140)]
```

# 문제7

```python
from numpy.ma.core import reshape
import numpy as np

a = [[4,3,2,1],[1,3,2,4],[7,30,15,18],[10,19,12,18]]
b = [[3,4,2,1],[2,3,4,4],[7,6,5,8],[7,9,2,8]]

aa = np.array(a)
bb = np.array(b)

# 행렬 aa와 bb의 행렬곱을 구하시오.
r = np.dot(aa,bb)
print(r,"\n")  # 출력 결과 : [[ 39  46  32  40] [ 51  61  32  61] [312 370 245 391] [278 331 192 326]]

# 행렬 bb와 aa의 행렬곱을 구하시오.
r1 = np.dot(bb,aa)
print(r1,"\n")  # 출력 결과 : [[ 40 100  56  73] [ 79 211 118 158] [149 341 197 265] [131 260 158 223]]

# 1에서 100까지 숫자중 3의 배수의 array을 생성해서 cc에 넣으시오
############## 다른 풀이 찾아보기 ##############
cc = np.arange(3,101,3)
print(cc,"\n")   # 출력 결과 : [ 3  6  9 12 15 18 21 24 27 30 33 36 39 42 45 48 51 54 57 60 63 66 69 72 75 78 81 84 87 90 93 96 99]

# 행렬 bb에서 4미만인 값만 3배하시오.
bb[bb<4] *= 3
print(bb,"\n")   # 출력 결과 : [[9 4 6 3] [6 9 4 4] [7 6 5 8] [7 9 6 8]]

# 자료형을 float로 변환하고, 2x8 행으로 변경

aa = aa.astype("float")
aa = aa.reshape(2,8)


print(aa)   # 출력 결과 : [[ 4.  3.  2.  1.  1.  3.  2.  4.] [ 7. 30. 15. 18. 10. 19. 12. 18.]]

```
