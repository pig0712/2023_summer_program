## 3. DataFrame의 1997년 11월의 'avgTa', 'minTa', 'maxTa' 컬럼의 데이터를 시각화 하기
- x축은 index가 오도록 하세요

```python
df[(df["iscs"].isnull() != True) & (df["iscs"].str.contains("비"))]
```
